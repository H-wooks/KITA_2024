## PyTorch vs. Tensorflow/Keras

### 1. **PyTorch**

#### 1.1. **개요**
PyTorch는 Facebook의 AI Research Lab에서 개발한 오픈 소스 딥러닝 프레임워크입니다. Pythonic한 인터페이스와 동적 계산 그래프(dynamic computational graph) 특성 덕분에 학습 곡선이 상대적으로 완만하고, Python 스타일의 코드를 사용할 수 있어서 자연스럽게 익숙해질 수 있습니다. 주로 연구 커뮤니티에서 인기를 얻었고, 최근에는 산업적 사용도 점점 증가하고 있습니다.

#### 1.2. **주요 특징**
- **동적 계산 그래프 (Dynamic Computation Graph)**
  - PyTorch의 가장 큰 특징 중 하나는 동적 계산 그래프입니다. 이를 통해 런타임에 그래프가 생성되므로, 코드 디버깅이나 변경이 훨씬 용이합니다. 이 특성 덕분에 PyTorch는 특히 연구에서 새로운 아이디어를 실험하기에 적합합니다.
  
- **직관적인 인터페이스**
  - Pythonic한 코딩 스타일을 가지고 있으며, 익숙한 Python 문법을 사용합니다. 이로 인해 TensorFlow보다 코드 작성이 상대적으로 간결하고 직관적입니다.
  
- **TorchScript**
  - PyTorch는 TorchScript를 사용하여 Python과 독립적으로 모델을 실행할 수 있게 해줍니다. 이는 모델을 실제 프로덕션 환경에 배포할 때 유용합니다.
  
- **Autograd**
  - PyTorch의 Autograd 기능은 자동 미분을 처리하는데, 이를 통해 백프로퍼게이션 과정에서 파생되는 모든 미분을 자동으로 계산할 수 있습니다.

- **커뮤니티**
  - 연구 커뮤니티와 긴밀하게 연계되어 있어 많은 논문에서 PyTorch로 구현된 코드가 제공됩니다. 딥러닝 연구에 적합한 프레임워크로 자리 잡고 있습니다.

### 2. **TensorFlow / Keras**

#### 2.1. **개요**
TensorFlow는 Google Brain 팀에서 개발한 오픈 소스 딥러닝 프레임워크로, Keras는 TensorFlow와 통합되어 사용되는 고수준 API입니다. TensorFlow는 생산 환경에서 대규모로 딥러닝 모델을 구축하고 배포하기에 적합하며, Keras는 사용자 친화적인 인터페이스를 제공하여 초보자들도 쉽게 사용할 수 있도록 설계되었습니다.

#### 2.2. **주요 특징**
- **정적 계산 그래프 (Static Computation Graph)**
  - TensorFlow는 초기에는 정적 계산 그래프를 사용했습니다. 즉, 그래프를 먼저 정의하고 나중에 실행하는 방식으로, 한 번 그래프를 정의하면 고정되어 변형이 어렵습니다. 최근에는 TensorFlow 2.0 이후 동적 계산 그래프(Eager Execution)를 도입해 PyTorch처럼 동적 그래프를 지원합니다.
  
- **확장성과 배포**
  - TensorFlow는 대규모 배포 환경에서 매우 강력합니다. TensorFlow Serving, TensorFlow Lite, TensorFlow.js 등 다양한 배포 옵션을 제공하여 모바일, 웹, 임베디드 디바이스까지 모델을 배포할 수 있습니다.
  
- **Keras (TensorFlow의 고수준 API)**
  - Keras는 TensorFlow와 통합된 고수준 API로, 사용자 친화적이고 빠르게 모델을 설계, 구축, 학습할 수 있습니다. 특히 Keras는 초보자들이 쉽게 딥러닝 모델을 시작할 수 있게 해줍니다.
  
- **TensorBoard**
  - TensorFlow는 강력한 시각화 도구인 TensorBoard를 제공합니다. 이를 통해 모델의 학습 과정을 실시간으로 모니터링하고, 그래프 및 파라미터의 변화를 시각적으로 확인할 수 있습니다.

- **TensorFlow Hub**
  - TensorFlow Hub는 사전 학습된 모델을 재사용할 수 있는 플랫폼으로, 연구자나 개발자들이 이미 학습된 모델을 쉽게 가져와 사용하거나 재학습할 수 있게 도와줍니다.

### 3. **PyTorch와 TensorFlow/Keras의 비교**

| **비교 항목**           | **PyTorch**                           | **TensorFlow/Keras**                |
|------------------------|---------------------------------------|-------------------------------------|
| **계산 그래프**         | 동적 계산 그래프 (Dynamic Graph)       | 정적 계산 그래프 (Static Graph) + 동적 지원 (Eager Execution) |
| **코딩 스타일**         | Pythonic, 간결하고 직관적             | Keras: 간결, TensorFlow: 다소 복잡함 |
| **배포 및 확장성**      | TorchScript로 배포 가능                | TensorFlow Serving, Lite, JS 등 다양한 배포 옵션 |
| **자동 미분 (Autograd)**| 매우 직관적                            | 자동 미분 제공하지만, PyTorch만큼 직관적이지 않음 |
| **성능**                | 연구 중심, 모델 실험에 적합            | 대규모 배포 환경에서 강력한 성능 제공 |
| **커뮤니티**            | 연구 중심, 최신 논문 구현 많이 제공    | 산업계와 연구계 모두 널리 사용됨    |
| **시각화 도구**         | TensorBoard 대안은 부족함              | TensorBoard 제공, 강력한 시각화 도구 |
| **모바일 및 웹 지원**   | 비교적 미약한 지원                    | TensorFlow Lite, TensorFlow.js 등 폭넓은 지원 |
| **프로덕션 환경 지원**  | TorchServe로 배포 가능하지만 한정적     | TensorFlow Extended(TFX)로 대규모 배포 지원 |

### 4. **결론**

- **PyTorch**는 직관적인 코드 작성과 동적 계산 그래프 덕분에 **연구 및 실험**에 적합합니다. 새로운 아이디어를 빠르게 실험하고 싶거나, 딥러닝 모델을 배우는 데 자연스럽게 접근하고 싶다면 PyTorch가 좋은 선택입니다.
- **TensorFlow/Keras**는 **대규모 배포**와 **프로덕션 환경**에서 딥러닝 모델을 사용하는 데 더 적합합니다. 특히 TensorFlow는 다양한 배포 옵션과 확장성을 지원하며, Keras는 간결한 API로 초보자들이 쉽게 접근할 수 있는 장점이 있습니다.

둘 다 매우 강력한 딥러닝 프레임워크로, 연구 목적인지, 대규모 배포 목적에 맞는지에 따라 선택할 수 있습니다.

PyTorch의 **동적 계산 그래프 (Dynamic Computation Graph)**는 딥러닝 모델을 구축하고 학습할 때 PyTorch가 제공하는 매우 중요한 기능 중 하나입니다. 이는 "즉시 실행 (eager execution)"이라고도 불리며, TensorFlow와 같은 프레임워크의 정적 계산 그래프(static computation graph)와 대비됩니다.

### 동적 계산 그래프의 핵심 개념
동적 계산 그래프는 학습 중에 그래프를 즉시 정의하고 실행하는 방식입니다. 즉, 계산을 수행할 때마다 새로운 그래프를 만들고, 이를 바로 실행하는 특징을 갖고 있습니다. 이를 통해 계산 과정 중 변수나 모델 구조의 변경이 매우 유연하게 이루어질 수 있습니다.

#### 1. **즉시 실행 (Eager Execution)**
PyTorch는 코드가 실행될 때마다 계산 그래프를 동적으로 생성하고, 이를 즉시 실행합니다. 즉, 딥러닝 모델에서 데이터를 전방 패스로 전달할 때, 해당 연산이 바로 수행되고 그 결과가 즉시 반환됩니다. 이런 즉시 실행 방식은 디버깅을 매우 용이하게 만들어 줍니다.

예를 들어, 다음과 같은 코드를 생각해 봅시다:

```python
import torch

x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)

z = x * y + y
```

여기서 `z`를 계산할 때 `x * y + y`가 바로 실행되고, 이 과정에서 계산 그래프가 동적으로 생성됩니다. 이후 `z.backward()`를 호출하면, PyTorch는 자동으로 그래프를 통해 미분을 수행하여 각 변수의 gradient를 계산합니다.

#### 2. **유연성 (Flexibility)**
동적 계산 그래프는 모델 구조를 즉석에서 정의하고, 계산 과정 중에도 그래프의 변경이 가능합니다. 이는 복잡한 구조를 가진 모델을 구현할 때 매우 유용합니다. 예를 들어, if문이나 for문을 통해 네트워크의 구조를 유연하게 변경할 수 있습니다.

```python
for i in range(5):
    if i % 2 == 0:
        z = z * x
    else:
        z = z + y
```

이 코드는 실행할 때마다 그래프의 구조가 바뀔 수 있는데, PyTorch의 동적 계산 그래프는 이러한 경우에도 문제없이 작동합니다. 이는 정적 계산 그래프에서 불가능하거나 복잡한 코드로 구현해야 하는 것을 동적 계산 그래프에서는 자연스럽게 구현할 수 있게 해줍니다.

#### 3. **Autograd: 자동 미분 기능**
PyTorch에서 동적 계산 그래프는 **Autograd** 기능과 밀접하게 연결되어 있습니다. Autograd는 계산 그래프를 추적하여, 미분이 필요한 연산에 대한 그래프를 자동으로 만들어 줍니다. 연산이 수행되면 PyTorch는 이를 기록하고, `backward()` 함수를 호출할 때 그래프를 따라 역전파를 통해 기울기(gradient)를 계산합니다.

```python
z.backward()  # z의 연산을 바탕으로 역전파 수행
print(x.grad)  # x의 gradient 값 출력
```

Autograd는 동적으로 그래프를 구성하기 때문에, 계산 그래프는 전방 패스(forward pass) 중에 실시간으로 생성되고, 역전파(backward pass)가 끝나면 필요에 따라 버려집니다. 이는 메모리 효율성을 높이고, 메모리 관리를 쉽게 해줍니다.

### 동적 계산 그래프의 장점
- **디버깅이 용이함**: 계산이 즉시 실행되므로, 중간 결과를 쉽게 확인할 수 있어 디버깅 과정이 매우 직관적입니다. 오류가 발생하면 즉시 알 수 있습니다.
- **유연한 모델 설계**: 반복문, 조건문 등을 사용해 네트워크 구조를 동적으로 변경할 수 있으며, 이는 매우 복잡한 모델을 쉽게 구현하는 데 도움을 줍니다.
- **학습 시 효율적**: 필요할 때만 그래프가 생성되고 역전파에 사용되며, 이후 메모리에서 제거되므로 메모리 사용이 효율적입니다.

### 정적 계산 그래프와의 차이점
- **정적 계산 그래프 (TensorFlow, Theano)**는 모델을 먼저 정의한 후, 그래프 전체를 컴파일하여 실행합니다. 이런 방식은 실행 전에 최적화된 모델을 만들 수 있지만, 그래프를 수정하려면 다시 컴파일해야 하므로 덜 유연합니다.
- 반면, **동적 계산 그래프 (PyTorch)**는 실행 시점에서 그래프를 생성하므로 더 유연하지만, 실행 속도는 정적 그래프에 비해 다소 느릴 수 있습니다. 하지만 최근에는 PyTorch에서도 정적 그래프처럼 최적화를 할 수 있는 방법이 연구되고 있습니다(예: TorchScript).

### PyTorch의 동적 계산 그래프 작동 방식
PyTorch의 동적 계산 그래프는 연산을 할 때마다 다음과 같은 과정을 거칩니다:

1. **Tensor와 연산 정의**: 계산 그래프는 Tensor와 연산으로 구성됩니다. Tensor는 `requires_grad=True` 옵션을 통해 그라디언트를 추적할지 여부를 설정할 수 있습니다.
2. **전방 패스 실행**: 전방 패스가 실행되면서 연산 결과가 생성되고, 그와 동시에 계산 그래프가 생성됩니다.
3. **역전파 실행**: `backward()` 호출 시, 계산 그래프를 타고 내려가면서 그라디언트가 자동으로 계산됩니다.

결론적으로, PyTorch의 동적 계산 그래프는 코드가 실행될 때마다 실시간으로 그래프를 구성하여 유연한 모델 설계와 디버깅을 가능하게 합니다. 이 특성 덕분에 PyTorch는 연구나 개발 환경에서 널리 사용되며, 특히 복잡하고 유연한 모델 설계에 적합한 딥러닝 프레임워크로 자리 잡았습니다.

### 텐서는 기본적으로 다차원 배열. 숫자들의 집합을 하나의 데이터 구조로 묶은 것인데, 이 구조가 단순한 숫자 리스트부터 매우 복잡한 다차원 배열까지 다양하게 지원; 좀 더 구체적으로 설명하자면:

- 스칼라 (0차원 텐서): 그냥 하나의 숫자입니다. 예를 들어 3이나 -5.2 같은 숫자들이 이에 해당합니다.
벡터 (1차원 텐서):

- 숫자들의 리스트로, 1차원 배열이라고 할 수 있습니다. 예를 들어 [1, 2, 3]이나 [-7, 4, 0] 같은 것들입니다.

- 행렬 (2차원 텐서):
행과 열로 이루어진 2차원 배열입니다. 예를 들어, 2x2 행렬은 다음과 같습니다:

[[1, 2],
 [3, 4]]

- 다차원 배열 (3차원 이상의 텐서):
3차원 이상의 배열로, 3차원 텐서는 여러 개의 행렬이 쌓여 있는 형태입니다. 예를 들어, 여러 장의 이미지 데이터를 3차원 텐서로 표현할 수 있습니다.

````python
## 여기서 z를 계산할 때 x * y + y 가 바로 실행되고, 이과정에서 계산 그래프가 동적으로 생성
## 이후 z.backward()를 호출하면 pytorch는 자동으로 그래프를 통해 미분을 수행하여 각 변수의 gradient를 계산
import torch

x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)

z = x * y + y
````

````python
## 동적 계산 그래프는 모델 구조를 즉성에서 정의하고, 계산 과정 중에도 그래프의 변경이 가능
## if문이나 for 문을 통해 네트워크의 구조를 유연하게 변경할 수 있습니다.
for i in range(5):
    if i % 2 == 0:
        z = z * x
    else:
        z = z + y
````

````python
## autograd는 계산 그래프를 추적하여, 미분이 필요한 연산에 대한 그래프를 자동으로 만들어 줍니다.
## 연산이 수행되면 PyTorch는 이를 기록하고, backward() 함수를 호출할 때 그래프를 따라 역전파를 통해 기울기(gradient)를 계산
z.backward()    # z의 연산을 바탕으로 역전파 수행
print(x.grad)   # x의 gradient를 출력
````

계산 그래프의 사례를 간단한 수식과 함께 살펴볼 수 있습니다. 예를 들어, 다음과 같은 수식을 고려해 보겠습니다:

\[
z = (x + y) * w
\]

여기서 \(x\), \(y\), \(w\)는 입력 변수이고, \(z\)는 결과입니다. 이 수식을 계산 그래프로 나타내면, 각 연산이 그래프의 노드로 표현되고, 변수와 연산 사이의 의존 관계가 엣지로 연결됩니다. 이제 이 수식을 계산 그래프로 표현한 사례를 보여드리겠습니다:

### 수식: \( z = (x + y) * w \)

1. **입력 노드**: \(x\), \(y\), \(w\)는 그래프의 입력 변수입니다.
2. **덧셈 연산 노드**: 먼저, \(x\)와 \(y\)를 더하는 연산이 일어납니다.
3. **곱셈 연산 노드**: 그다음, 덧셈 결과와 \(w\)를 곱하는 연산이 일어납니다.
4. **출력 노드**: 최종 결과인 \(z\)가 나옵니다.

### 계산 그래프의 시각적 표현
다음과 같이 계산 과정을 그래프로 나타낼 수 있습니다:

```
   x -----
          |
          + ----> (x + y) -----> * ----> z
          |                      |
   y -----                       w
```

여기서 각 단계는 다음과 같습니다:
- **덧셈 연산**: \(x + y\)는 하나의 노드로 표현됩니다.
- **곱셈 연산**: 그다음 덧셈 결과와 \(w\)의 곱이 또 다른 노드로 표현됩니다.
- **출력**: 최종 결과인 \(z\)가 마지막에 나옵니다.

### 역전파 과정
이 계산 그래프는 **역전파** 과정에서 중요한 역할을 합니다. 예를 들어, 손실 함수에서 \(z\)에 대한 기울기를 계산할 때, 연쇄 법칙에 따라 각 연산에 대한 기울기를 차례대로 계산하여 전달합니다.

1. **\(z\)에 대한 기울기 계산**: 먼저, \(z\)에 대한 손실 함수의 기울기를 구합니다.
2. **곱셈 연산의 기울기**: \(z\)의 기울기를 \(w\)와 \((x + y)\) 각각에 대해 계산하여 전달합니다.
3. **덧셈 연산의 기울기**: \((x + y)\)의 기울기를 \(x\)와 \(y\)에 대해 계산하여 전달합니다.

이 과정에서 PyTorch와 같은 프레임워크는 자동으로 역전파를 수행해 각 파라미터에 대한 기울기를 구할 수 있습니다.

### 더 복잡한 사례
실제 딥러닝 모델에서는 이보다 훨씬 복잡한 계산 그래프가 형성됩니다. 예를 들어, **합성곱 신경망(CNN)**이나 **순환 신경망(RNN)**은 여러 층(layer)과 비선형 활성화 함수, 매트릭스 연산 등이 포함되어 있으며, 이러한 모든 연산이 하나의 거대한 계산 그래프로 연결됩니다. 각 층의 가중치와 입력 간의 연산들이 일련의 계산 그래프를 형성하고, 이 그래프를 통해 역전파가 이루어집니다.

## 텐서(Tensors)

* 데이터 표현을 위한 기본 구조로 텐서(tensor)를 사용
* 텐서는 데이터를 담기위한 컨테이너(container)로서 일반적으로 수치형 데이터를 저장
* 넘파이(NumPy)의 ndarray와 유사
* GPU를 사용한 연산 가속 가능

[ ndarray와 텐서 비교 ]
- GPU 지원: PyTorch의 Tensor와 TensorFlow의 텐서는 GPU를 통한 가속을 지원. 이는 대규모 배열 연산에서 상당한 성능 향상을 가져올 수 있는 반면, NumPy ndarray는 기본적으로 CPU에서만 작동(sklearn은 ndarray.. CUP에서만...)
- 자동 미분 지원: PyTorch와 TensorFlow는 딥러닝 모델 학습을 위해 자동 미분을 지원하는 반면, NumPy는 이러한 기능을 내장하고 있지 않다. 이는 신경망의 역전파와 같은 복잡한 연산을 구현할 때 중요한 차이점이다.
- API 호환성: PyTorch와 TensorFlow/Keras는 NumPy와의 호환성을 중시하며, 많은 경우에 NumPy의 API를 모방. 이는 NumPy 사용자가 쉽게 이러한 프레임워크로 전환할 수 있도록 돕는다. 그러나 각 프레임워크는 자체적인 최적화와 기능을 제공하기 때문에, 모든 NumPy 연산이 그대로 사용될 수 있는 것은 아니다.

[ PyTorch와 TensorFlow/Keras에서의 ndarray 사용 ]
- PyTorch에서의 사용: PyTorch는 torch.from_numpy() 함수를 통해 NumPy ndarray를 PyTorch의 Tensor로 변환할 수 있다. 변환된 텐서는 원래 ndarray와 메모리를 공유하기 때문에, 한 객체의 변경이 다른 객체에도 반영된다. 반대로, .numpy() 메소드를 사용하여 PyTorch 텐서를 NumPy 배열로 변환할 수 있다.
- TensorFlow/Keras에서의 사용: TensorFlow에서는 tf.convert_to_tensor() 함수를 사용하여 NumPy ndarray를 TensorFlow의 텐서로 변환할 수 있다. 이 변환된 텐서는 TensorFlow 연산에 사용될 수 있다. TensorFlow 텐서를 NumPy 배열로 변환하기 위해서는 .numpy() 메소드를 사용할 수 있다.

PyTorch와 TensorFlow/Keras 모두 NumPy ndarray와의 상호 운용성을 지원하며, 이를 통해 각 프레임워크의 텐서로 쉽게 변환할 수 있다. 각 프레임워크의 텐서는 GPU 지원과 자동 미분과 같은 추가적인 기능을 제공하여 딥러닝 모델의 개발과 학습을 효율적으로 할 수 있게 한다.

1. torch.rand() : 0과 1 사이의 숫자를 균등하게 생성

3. torch.randn() : 평균이 0이고 표준편차가 1인 가우시안 정규분포를 이용해 생성

7. torch.randperm() : 주어진 범위 내의 정수를 랜덤하게 생성

````python
import torch

# create a tensor with random values
tensor_rand=torch.rand(3,3) # 0~1

existing_tensor=torch.tensor([[1,2],[3,4]])
tensor_rand_like=torch.rand_like(existing_tensor, dtype=torch.float)

tensor_randn=torch.randn(3,3)
tensor_randn_like=torch.randn_like(existing_tensor, dtype=torch.float)        ## 구조는 input의 existing_tensor 형태 유지하면서 0~1까지의 수

tensor_randint=torch.randint(low=0, high=10, size=(3,3))
tensor_randint_like=torch.randint_like(existing_tensor, low=0, high=10)       ## 구조는 input의 existing_tensor 형태 유지하면서 0~9까지의 수

tensor_randperm=torch.randperm(10)    # 0부터 n-1까지의 정수를 무작위로 섞은 순열(permutation)을 반환, 중복없는 무작위 정수 시퀀스

print("tensor_rand\n", tensor_rand,"\n")
print("existing_tensor\n", existing_tensor,"\n")
print("tensor_rand_like\n",tensor_rand_like,"\n")
print("tensor_randint\n",tensor_randint,"\n")
print("tensor_randint_like\n",tensor_randint_like,"\n")
print("tensor_randperm\n",tensor_randperm,"\n")
````

squeeze() 함수는 크기가 1인 입력 텐서의 모든 차원을 제거하는 데 사용
- dim이 제공되지 않으면 squeeze()는 텐서에서 크기 1의 모든 차원을 제거
- dim이 제공되면 크기가 1인 경우에만 지정된 차원만 제거. 즉 (1, 2, 1, 3)인 텐서가 있고 dim이 2이면 결과의 크기는 (1, 2, 3)

````python
tensor=torch.rand(1,3,3)
print(tensor)
print(tensor.shape)
t=tensor.squeeze()
print(t)
print(t.shape)
````

tensor.unsqueeze(dim=2)

- unsqueeze 함수는 텐서에 새로운 차원을 추가. 여기서 dim=2는 새로운 차원이 추가될 위치를 나타내며 기존 텐서의 모든 차원의 인덱스를 증가시키고, 새로운 차원은 dim=2에 위치.

- 원래 텐서의 모양(shape)이 (3, 4)이었다면, unsqueeze(dim=2)를 적용한 후의 모양은 (3, 4, 1)이 됩니다. 따라서, 이 코드에서는 기존 텐서에 1차원을 추가

````python
t=torch.rand(3,3)
print(t)
print(t.shape)
tensor=t.unsqueeze(dim=0)
print(tensor)
print(tensor.shape)
````

stack은 새로운 차원을 추가하여 텐서들을 결합하는 반면, cat은 기존 차원을 따라 텐서들을 연결

- stack : torch.stack()은 새로운 차원을 추가한 후 텐서를 결합합니다. 즉, 텐서들의 차원 수가 1 증가하며, 지정된 차원에 새로운 차원을 추가한 뒤 해당 차원을 따라 텐서들을 쌓습니다.
- cat: torch.cat()은 기존의 차원을 따라 텐서들을 연결합니다. 즉, 텐서의 차원 수는 그대로 유지되며, 지정된 차원(axis)에서 텐서들이 연결됩니다.

````python
import torch

# 두 개의 텐서 생성
x = torch.FloatTensor([1, 4])
print(x)
y = torch.FloatTensor([2, 5])
print(y)
z = torch.FloatTensor([3, 6])
print(z)
print()
print(torch.stack([x, y, z]))
````

````python
# torch.cat : dim = 0 이므로 1 + 1 =  2가 됨
a = torch.randn(1, 3, 3)
print(a, '\n')
b = torch.randn(1, 3, 3)
print(b, '\n')
c = torch.cat((a, b), dim=0)
print(c, '\n')
print(c.size())
````

###chunk
chunk: torch.chunk()는 텐서를 주어진 개수의 균등한 크기로 나눕니다. 나누어진 텐서들의 크기는 가능한 한 동일하게 유지되며, 전체 크기를 나누어 떨어지지 않으면 마지막 텐서는 더 작은 크기가 될 수 있습니다.(몇 개로 나눌 것인가?)

split: torch.split()은 텐서를 주어진 크기를 기준으로 나눕니다. 분할되는 각 텐서의 크기를 직접 지정할 수 있습니다. 크기가 균등하지 않아도 되며, 마지막 텐서의 크기는 자동으로 조정됩니다. (텐서의 크기는 몇인가?)

````python
# chunk
tensor = torch.rand(3, 6)
print(tensor)

t1, t2 = torch.chunk(tensor, 2, dim=1)
print(t1, t1.shape)
print(t2, t2.shape)
````

````python
# split
tensor = torch.rand(3, 6)
t1, t2, t3 = torch.split(tensor, 2, dim=1)
print(t1, t1.shape)
print(t2, t2.shape)
print(t3, t3.shape)
print()

# 만약 dim = 0일 경우
tensor = torch.rand(3, 6)
t1, t2, t3 = torch.split(tensor, 1, dim=0)
print(t1, t1.shape)
print(t2, t2.shape)
print(t3, t3.shape)
````

### PyTorch 데이터 타입
PyTorch에서 Tensor는 다양한 데이터 타입을 가질 수 있으며, 각 데이터 타입은 Tensor의 저장 및 연산 방식에 영향을 미칩니다. PyTorch에서 지원하는 주요 데이터 타입은 각기 다른 크기와 표현 방식으로 데이터를 처리하는 데 사용되며, 특히 GPU와 CPU 간의 효율적인 메모리 사용과 성능 최적화에 중요합니다.

아래는 PyTorch에서 지원하는 주요 데이터 타입들을 정리한 표입니다.

| 데이터 타입                | 설명                              | 약어  |
|----------------------------|-----------------------------------|-------|
| `torch.float32` (기본값)    | 32비트 부동 소수점                | `torch.float` 또는 `torch.FloatTensor` |
| `torch.float64`             | 64비트 부동 소수점 (double precision) | `torch.double` 또는 `torch.DoubleTensor` |
| `torch.float16`             | 16비트 부동 소수점 (half precision) | `torch.half` 또는 `torch.HalfTensor` |
| `torch.int8`                | 8비트 정수                       | `torch.int8` |
| `torch.uint8`               | 8비트 부호 없는 정수             | `torch.uint8` |
| `torch.int16`               | 16비트 정수                      | `torch.short` 또는 `torch.ShortTensor` |
| `torch.int32`               | 32비트 정수                      | `torch.int` 또는 `torch.IntTensor` |

### 데이터 타입 설명

1. **부동 소수점 타입 (Floating-Point Types)**
   - `torch.float32` (기본값): 32비트 부동 소수점 타입입니다. 딥러닝에서 일반적으로 사용되는 데이터 타입이며, 많은 연산의 기본 데이터 타입입니다.
   - `torch.float64`: 64비트 부동 소수점 타입으로, 더 높은 정밀도를 필요로 하는 연산에 사용됩니다. 메모리 사용량은 `torch.float32`에 비해 2배이지만, 연산 속도는 느릴 수 있습니다.
   - `torch.float16`: 16비트 부동 소수점 타입으로, 메모리 사용량을 줄이고 빠른 연산을 위해 사용됩니다. 주로 GPU에서 사용되며, 정밀도가 낮아 질 수 있지만, 최근의 GPU들은 이를 효율적으로 처리합니다.
   - `torch.bfloat16`: 16비트 블록 부동 소수점 타입으로, 부동 소수점보다 적은 비트로 표현하지만, 적은 정밀도로도 더 넓은 범위를 표현할 수 있습니다. 주로 메모리 최적화와 성능 향상 목적으로 사용됩니다.

2. **정수 타입 (Integer Types)**
   - `torch.int8`: 8비트 정수 타입으로, 매우 작은 메모리 공간을 차지하는 정수 연산에 사용됩니다.
   - `torch.uint8`: 8비트 부호 없는 정수 타입으로, 음수를 표현하지 않고 0에서 255 범위의 양수만 표현하는 정수형입니다. 주로 이미지 데이터에서 픽셀 값으로 사용됩니다.
   - `torch.int16`: 16비트 정수 타입으로, 음수를 포함한 정수 표현이 가능합니다.
   - `torch.int32`: 32비트 정수 타입으로, 딥러닝에서 주로 사용되는 정수 타입입니다.
   - `torch.int64`: 64비트 정수 타입으로, 더 큰 범위의 정수를 필요로 하는 연산에 사용됩니다. PyTorch에서는 `torch.long`이 기본적으로 `torch.int64`를 가리킵니다.

3. **Boolean 타입**
   - `torch.bool`: Boolean 타입으로, 참(True)과 거짓(False) 값만을 가집니다. 논리 연산에 사용되며, 메모리 효율적입니다.

### 데이터 타입 선택 가이드
- **`float32`**: 딥러닝에서 기본적으로 사용되는 타입. 대부분의 모델에서 연산 성능과 메모리 사용량의 균형이 잘 맞음.
- **`float64`**: 높은 정밀도가 필요한 계산에서 사용. 연산 성능은 떨어지지만, 정밀도가 중요할 때 선택.
- **`float16`** 및 **`bfloat16`**: 메모리 절약과 속도를 위해 주로 GPU 환경에서 사용. 특히 큰 모델에서 메모리 부족 문제를 해결하는 데 유용.
- **`int32` 및 `int64`**: 정수 계산에서 사용. 기본적으로 `int64`는 큰 정수를 다룰 수 있지만 메모리 사용량이 크므로, `int32`를 사용할 수 있는 경우 선택.
- **`bool`**: 논리 연산 및 마스크 연산에 사용. True/False를 표현하는 데 매우 효율적.

#### 텐서를 GPU에서 CPU로 보내는 이유는 몇 가지가 있을 수 있습니다:

- 호환성 및 디바이스 제약:

- 모델 배포: 최종 모델을 배포하거나 다른 환경에서 실행할 때 GPU가 없는 시스템에서 사용할 수 있도록 CPU로 텐서를 이동시킵니다.
장치 간 호환성: 어떤 작업들은 GPU에서 지원되지 않거나 제대로 동작하지 않을 수 있습니다. 이 경우 CPU로 이동시켜 작업을 수행해야 합니다.
디버깅:

- 디버깅 및 로깅: GPU 텐서를 직접 출력하거나 디버깅하는 것이 어려울 수 있습니다. CPU 텐서는 출력이 더 직관적이고, 디버깅 도구와의 호환성이 좋습니다.
저장 및 직렬화:

- 모델 저장: 텐서나 모델을 저장하기 전에 CPU로 이동시키는 것이 일반적입니다. GPU 텐서를 그대로 저장하면, GPU가 없는 시스템에서 모델을 불러올 때 오류가 발생할 수 있습니다.
데이터 직렬화: 텐서를 직렬화해서 파일로 저장하거나 네트워크를 통해 전송할 때, CPU 텐서로 변환하면 호환성 문제가 줄어듭니다.
후처리 및 시각화:

- 데이터 후처리: 시각화 라이브러리나 특정 후처리 도구들이 CPU에서만 동작하는 경우가 많습니다. 예를 들어, Matplotlib로 그래프를 그릴 때, GPU 텐서를 CPU로 이동시켜야 올바르게 처리할 수 있습니다.
메모리 관리:

- 메모리 관리: GPU 메모리는 한정되어 있기 때문에, 계산이 끝난 후 더 이상 GPU 메모리를 점유하지 않도록 CPU로 텐서를 이동시킬 수 있습니다.

- 텐서를 numpy 배열로 변환하기 위해서도 CPU로 이동시키는 것이 필요 GPU에 있는 텐서는 직접적으로 numpy 배열로 변환할 수 없기 때문에, 먼저 CPU로 이동
  - a = a.to('cpu') 또는 a.cpu()
  - numpy() 메서드를 사용하여 numpy 배열로 변환할 수 있습니다.
예를 들어, a_numpy = a.cpu().numpy()

##### 다음 매개변수들을 사용하여 FashionMNIST 데이터셋 을 불러옵니다:
- root 는 학습/테스트 데이터가 저장되는 경로입니다.
- train 은 학습용 또는 테스트용 데이터셋 여부를 지정합니다.
- download=True 는 root 에 데이터가 없는 경우 인터넷에서 다운로드합니다.
- transform 과 target_transform 은 특징(feature)과 정답(label) 변형(transform)을 지정합니다.

````python
## 공개 데이터셋에서 학습 데이터를 down
training_data=datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,        ## 데이터셋이 로컬에 없으면 자동으로 인터넷에서 다운로드
    transform=ToTensor(), ## 이미지 데이터를 PyTorch 텐서로 변환; 0255의 픽셀 값--> 0/1 사이의 값으로 정규화하고 텐서 형태로 변환, 모델에 입력할 수 있게 변환
)

## 공개 데이터셋에서 테스트 데이터를 down
test_data=datasets.FashionMNIST(
    root="data",
    train=False,        ## 테스트 데이터
    download=True,      ## 이미 다운로드된 경우, 해당 디렉토리에서 데이터를 불러옵
    transform=ToTensor(),
)
````

### 데이터셋을 순회하고 시각화하기

Dataset 을 DataLoader 의 인자로 전달합니다. 이는 데이터셋을 순회 가능한 객체(iterable)로 감싸고, 자동화된 배치(batch), 샘플링(sampling),

즉, 데이터로더(dataloader) 객체의 각 요소는 64개의 특징(feature)과 정답(label)을 묶음(batch)으로 반환합니다.

````python
batch_size=64

# 데이터 로더를 생성
train_dataloader=DataLoader(training_data, batch_size=batch_size)
test_dataloader=DataLoader(test_data, batch_size=batch_size)

for X, y in test_dataloader:
    print("Shape of X [N, C, H, W]: ", X.shape)
    print("Shape of y: ", y.shape, y.dtype)
    break
````

- batch_size=64:batch_size는 한 번에 처리할 데이터의 양을 의미합니다. 여기서는 batch_size를 64로 설정했으므로, DataLoader가 한 번에 64개의 이미지와 그에 대응하는 레이블을 불러오게 됩니다.
DataLoader(training_data, batch_size=batch_size):

- DataLoader: PyTorch의 DataLoader는 데이터셋을 효율적으로 불러오는 도구입니다. 이 코드에서는 training_data와 test_data에서 데이터를 불러올 때 사용됩니다.
training_data: 학습 데이터셋을 사용합니다.
batch_size=batch_size: 한 번에 64개의 데이터를 로드합니다.
이 DataLoader는 학습 데이터를 64개씩 묶어서 모델에 공급할 수 있도록 준비합니다.

- for X, y in test_dataloader::
test_dataloader에서 데이터 배치를 반복적으로 가져옵니다. X는 이미지 배치이고, y는 그에 대응하는 레이블 배치입니다.
test_dataloader는 test_data에서 데이터를 로드하는 역할을 합니다. 마찬가지로, 배치 크기는 64입니다.
이 반복문은 test_dataloader에서 배치를 하나씩 가져와서 X와 y에 할당합니다.

- print("Shape of X [N, C, H, W]: ", X.shape):
X.shape는 배치된 이미지 데이터의 형상을 출력합니다.
[N, C, H, W]는 이미지 텐서의 형상을 설명하는데, 각각 다음을 의미합니다:
  - N: 배치 크기(batch size), 여기서는 64입니다.
  - C: 채널 수(channels), FashionMNIST는 흑백 이미지이므로 채널 수는 1입니다.
  - H: 이미지의 높이(height), FashionMNIST 이미지는 28픽셀입니다.
  - W: 이미지의 너비(width), 역시 28픽셀입니다.
따라서, 출력되는 형상은 [64, 1, 28, 28]이 됩니다.

- print("Shape of y: ", y.shape, y.dtype):
y.shape는 레이블 배치의 형상을 출력합니다.
y는 각 이미지에 해당하는 레이블을 담고 있으며, y.shape는 (batch_size,), 즉 [64]의 형상을 가집니다.
y.dtype는 레이블 데이터의 자료형(data type)을 출력합니다. 일반적으로 torch.int64가 됩니다.
- break:
이 break는 반복문을 한 번만 실행하고 종료하게 만듭니다. 따라서, 코드가 실행되면 첫 번째 배치의 형상만 출력됩니다.


###### 전체 동작 요약
DataLoader를 사용하여 학습 데이터와 테스트 데이터를 64개의 샘플로 이루어진 배치로 나누어 로드합니다.
테스트 데이터의 첫 번째 배치에 대해 이미지와 레이블의 형상을 출력합니다.
출력된 형상에서 X는 [64, 1, 28, 28]로, 64개의 흑백 이미지가 28x28 픽셀 크기로 배치되어 있음을 의미합니다. y는 [64]로, 64개의 레이블이 각각의 이미지에 대응합니다.
이 과정을 통해, 데이터를 모델에 공급하기 전에 배치 단위로 나누고, 그 형상을 확인하여 데이터가 올바르게 준비되었는지 확인할 수 있습니다.


PyTorch에서 딥러닝 모델을 구성하는 방법은 매우 직관적이며, 주요 방식은 클래스를 정의하여 모델을 설계하는 것입니다. PyTorch의 모든 모델은 torch.nn.Module 클래스를 기반으로 하며, 이 클래스를 상속하여 사용자 정의 모델을 구성할 수 있습니다. 아래는 PyTorch에서 모델을 구성하는 주요 단계와 방식에 대한 설명입니다.

### 1. 순전파 (Forward Propagation)
순전파는 입력 데이터를 모델에 전달해 예측값을 계산하는 과정입니다. 이때 PyTorch는 각 연산에 대한 계산 그래프를 내부적으로 생성하고, 이 그래프를 통해 역전파를 진행할 준비를 합니다.

```python
output = model(input_data)  # 순전파
```

### 2. 손실 함수 계산
순전파 결과인 예측값을 기반으로 손실(loss)을 계산합니다. 손실 함수는 예측값과 실제값 간의 차이를 측정하는 함수입니다.

```python
loss = loss_fn(output, target)  # 예측값과 실제값의 차이로 손실 계산
```

### 3. 역전파 (Backward Propagation)
역전파는 모델의 각 매개변수에 대한 기울기(gradient)를 계산하는 과정입니다. 손실 함수에 대한 각 매개변수의 편미분 값을 구하여 가중치를 업데이트할 수 있는 방향을 결정합니다. PyTorch에서는 `loss.backward()`를 호출함으로써 자동으로 역전파가 수행됩니다.

```python
loss.backward()  # 역전파 실행, 각 매개변수에 대한 기울기 계산
```

이 과정을 통해 **모델의 매개변수(가중치)**에 대해 손실 함수의 기울기가 계산되며, 계산된 기울기는 `parameter.grad` 속성에 저장됩니다.

### 4. 옵티마이저를 사용한 가중치 갱신
역전파를 통해 계산된 기울기를 바탕으로 옵티마이저(optimizer)가 가중치를 갱신합니다. 이를 위해 옵티마이저의 `step()` 함수를 호출합니다.

```python
optimizer.step()  # 기울기를 이용해 가중치 업데이트
```

### 역전파의 내부 동작
역전파가 어떻게 동작하는지를 자세히 설명하면 다음과 같습니다:

1. **계산 그래프 생성**: 순전파에서 각 연산이 수행될 때, PyTorch는 각 연산을 노드로 하는 계산 그래프를 생성합니다. 이 그래프는 텐서 간의 연산을 추적하며, 역전파를 위해 각 연산에 대한 미분 방법도 저장합니다.

2. **체인 룰 (Chain Rule)**: 역전파는 체인 룰(연쇄 법칙)을 이용하여 손실에 대한 각 매개변수의 기울기를 계산합니다. 체인 룰이란 복합 함수의 미분을 계산할 때, 각 함수의 미분을 곱하는 방법입니다. PyTorch는 이 체인 룰을 자동으로 적용해, 순전파에서 쌓인 연산들을 역순으로 따라가며 미분을 계산합니다.

3. **기울기 저장**: 각 연산을 역으로 추적하면서, 각 매개변수에 대해 손실 함수의 기울기를 계산하고, 이를 해당 매개변수의 `.grad` 속성에 저장합니다. 이 기울기는 이후 옵티마이저에 의해 사용되어 가중치가 갱신됩니다.

### 예시 코드

```python
import torch
import torch.nn as nn
import torch.optim as optim

# 간단한 선형 모델 정의
model = nn.Linear(2, 1)  # 입력 2, 출력 1

# 입력과 타겟 데이터 생성
input_data = torch.tensor([[1.0, 2.0]], requires_grad=True)
target = torch.tensor([[1.0]])

# 손실 함수와 옵티마이저 정의
loss_fn = nn.MSELoss()  # Mean Squared Error Loss
optimizer = optim.SGD(model.parameters(), lr=0.01)

# 순전파
output = model(input_data)
loss = loss_fn(output, target)

# 역전파
optimizer.zero_grad()  # 기울기 초기화
loss.backward()        # 역전파 실행
optimizer.step()       # 옵티마이저를 통한 가중치 갱신
```

### `requires_grad`와 계산 기록
PyTorch에서는 텐서의 `requires_grad` 속성을 `True`로 설정하면 해당 텐서와 관련된 모든 연산을 추적하여 자동 미분을 가능하게 합니다.

- 텐서 `input_data`에서 `requires_grad=True`로 설정했기 때문에, 역전파 시 해당 텐서에 대한 기울기도 계산됩니다.
- 계산 그래프를 추적하고 싶지 않은 경우, `torch.no_grad()` 블록 안에서 연산을 수행할 수 있습니다.

```python
with torch.no_grad():
    output = model(input_data)  # 역전파 없이 순전파만 수행
```

### 중요 개념 요약
1. **순전파**: 입력 데이터를 모델에 전달해 출력을 계산하고, 각 연산을 추적하여 계산 그래프를 만듭니다.
2. **손실 계산**: 예측값과 실제값의 차이를 계산하여 손실(loss)을 구합니다.
3. **역전파**: 손실 함수에 대한 각 매개변수의 기울기를 계산합니다. PyTorch는 자동으로 체인 룰을 적용해 기울기를 계산합니다.
4. **기울기 업데이트**: 옵티마이저가 계산된 기울기를 사용하여 매개변수(가중치)를 갱신합니다.

이 과정이 반복되면서 모델의 성능이 점차 개선되게 됩니다.

## 모델 만들기
PyTorch에서 신경망 모델은 nn.Module을 상속받는 클래스(class)를 생성하여 정의합니다. ``__init__`` 함수에서 신경망의 계층(layer)들을 정의하고 ``forward`` 함수에서 신경망에 데이터를 어떻게 전달할지 지정합니다. 가능한 경우 GPU로 신경망을 이동시켜 연산을 가속(accelerate)합니다.

#### 학습을 위한 장치 얻기
가능한 경우 GPU와 같은 하드웨어 가속기에서 모델을 학습하려고 합니다. torch.cuda가 사용 가능한지 확인해보고, 그렇지 않으면 CPU를 계속 사용합니다.

#### 클래스 정의하기
신경망 모델을 nn.Module 의 하위클래스로 정의하고, __init__ 에서 신경망 계층들을 초기화합니다. nn.Module 을 상속받은 모든 클래스는 forward 메소드에 입력 데이터에 대한 연산들을 구현합니다.

````python
# 학습에 사용할 CUP나 MPS 장치를 설정
device = (
    "cuda"
    if torch.cuda.is_available()
    # else "mps"
    # if torch.backends.mps.is_available()            ## mps는 Apple의 M1/M2 칩에서 제공하는 GPU 가속 기능
    else "cpu"
)
print(f"Using {device} device")

# 모델을 정의
class NeuralNetwork(nn.Module):     # NeuralNetwork 클래스: PyTorch의 신경망 모듈(nn.Module)을 상속하여 간단한 신경망을 정의
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()         ## 28x28 크기의 2D 이미지를 1D 벡터로 변환합니다. 예를 들어, [28, 28] 크기의 이미지를 [784] 크기의 벡터로 변환
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),          ## 입력 노드가 784(28x28)개이고 출력 노드가 512개인 선형 계층(fully connected layer)
            nn.ReLU(),
            nn.Linear(512, 512),            ## 두 번째 선형 계층으로, 512개의 입력을 받아 512개의 출력을 생성
            nn.ReLU(),
            nn.Linear(512, 10)              # 마지막 선형 계층, 512개의 입력을 받아 10개의 출력을 생성. 이 10개의 출력은 이미지가 10개 클래스 중 어떤 클래스에 속하는지를 나타냄
        )

    def forward(self, x):
        x = self.flatten(x)                 # 입력 이미지를 평평하게 만들어 1D 벡터로 변환
        logits = self.linear_relu_stack(x)  # 변환된 입력을 위에서 정의한 계층에 통과시켜 결과값(logits)을 얻습
        return logits

model = NeuralNetwork().to(device)          ## NeuralNetwork 클래스의 인스턴스를 생성한 후, 지정된 장치(device)로 모델을 이동
print(model)
````

##### 계층별 설명:
1. Flatten 계층 (1개):
이 계층은 28x28 크기의 2D 이미지 데이터를 1D 벡터(크기: 784)로 변환합니다. 이 Flatten 계층은 신경망이 2D 이미지 데이터를 fully connected 계층에서 처리할 수 있도록 평탄화합니다.

2. 첫 번째 Linear 계층 (2개):
입력 노드 수: 784 (Flatten 계층에서 나온 벡터)
출력 노드 수: 512
이 계층에서는 784개의 입력 노드가 512개의 출력 노드로 완전 연결됩니다.

3. 첫 번째 ReLU 계층 (3개):
ReLU (Rectified Linear Unit) 활성화 함수는 비선형성을 추가합니다. 이 계층은 첫 번째 Linear 계층의 출력을 받아, 비선형 활성화 함수를 적용합니다.

4. 두 번째 Linear 계층 (4개):
입력 노드 수: 512
출력 노드 수: 512
이 계층에서는 512개의 입력 노드가 512개의 출력 노드로 완전 연결됩니다.

5. 두 번째 ReLU 계층 (5개):
두 번째 Linear 계층의 출력에 비선형성을 추가합니다.

6. 세 번째 Linear 계층:
입력 노드 수: 512
출력 노드 수: 10

- 이 마지막 계층에서는 512개의 입력 노드가 10개의 출력 노드로 완전 연결됩니다. - 이 10개의 노드는 각 클래스(예: FashionMNIST의 10개 의류 카테고리)에 해당하는 출력입니다.

##### 전체 네트워크 구성:
- 이 모델은 Flatten 계층을 제외하면 3개의 Linear 계층과 2개의 ReLU 활성화 계층으로 구성
- 각 Linear 계층은 이전 계층의 모든 노드와 완전히 연결되어 있으며, 이들 사이에 ReLU 활성화 함수가 적용되어 비선형성을 추가합니다.
- 결과적으로, 이 모델은 28x28 크기의 이미지를 입력으로 받아 10개의 클래스 중 하나로 분류하는 fully connected 계층이 3개인 구조를 가집니다.
- Flatten 계층을 포함하면 모델에는 총 6개의 계층이 있지만, fully connected된 계층은 3개이며, ReLU 활성화 함수까지 포함한 계층 수는 총 5개입니다.

````python
##### 위 코드와 동일하지만 모두 manually 계층 설정한 코드 (위 코드와 실질적으로는 동일)


class NeuralNetwork(nn.Module):     # NeuralNetwork 클래스: PyTorch의 신경망 모듈(nn.Module)을 상속하여 간단한 신경망을 정의
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()         ## 28x28 크기의 2D 이미지를 1D 벡터로 변환합니다. 예를 들어, [28, 28] 크기의 이미지를 [784] 크기의 벡터로 변환
        self.fc1=nn.Linear(28*28, 512)      ## 첫번째 선형 계층
        self.relu1=nn.ReLU()                ## 첫번째 활성화 함수
        self.fc2=nn.Linear(512, 512)        ## 두번째 선형 계층
        self.relu2=nn.ReLU()                ## 두번째 활성화 함수
        self.fc3=nn.Linear(512, 10)         ## 세번째 선형 계층 (출력 레이어)

    def forward(self, x):
        x = self.flatten(x)                 ## 입력 이미지를 평탄화
        x=self.fc1(x)                       ## 첫번째 선형 계층 통과
        x=self.relu1(x)                     ## 첫번째 비선형 활성화 함수 계층 통과
        x=self.fc2(x)                       ## 두번째 선형 계층 통과
        x=self.relu2(x)                     ## 두번째 비선형 활성화 함수 계층 통과
        x=self.fc3(x)                       ## 세번째 선형 계층 통과 (출력)
        return x

## 모델을 정의하고 장치에 할당
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = NeuralNetwork().to(device)          ## NeuralNetwork 클래스의 인스턴스를 생성한 후, 지정된 장치(device)로 모델을 이동
print(model)
````

````python
#### 활성화 함수 참고용
import numpy as np
import matplotlib.pyplot as plt
from numpy import linspace

# Define functions
x = linspace(-5, 5, 100)
y1 = 1 / (1 + np.exp(-x))  # Sigmoid
y2 = np.tanh(x)            # tanh
y3 = np.maximum(0, x)      # LeLU
y4 = np.log(1 + np.exp(x)) # Softplus (used instead of softmax for this comparison)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x, y1, 'b', label='Sigmoid')
plt.plot(x, y2, 'r', label='tanh')
plt.plot(x, y3, 'g', label='LeLU')
plt.plot(x, y4, 'm', label='Softplus')
plt.legend()
plt.title("Comparison of Activation Functions")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.grid(True)
plt.show()
````

## 모델 매개 변수 최적화 하기
- 모델을 학습하려면 손실 함수 (loss function)와 옵티마이저(optimizer)가 필요

````python
loss_fn=nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(model.parameters(), lr=1e-3)


## nn.CrossEntropyLoss():
# 이 손실 함수는 다중 클래스 분류 문제에서 주로 사용됩니다.
# 모델의 출력(예측 값)과 실제 레이블(정답) 간의 차이를 측정하는 역할을 합니다.
# Cross-Entropy 손실 함수는 모델이 예측한 클래스 확률 분포와 실제 클래스 레이블 간의 불일치를 계산하여, 이를 최소화하도록 모델을 학습시킵니다.
# 이 함수는 내부적으로 Softmax 함수를 포함하고 있어, 모델의 출력 로짓(logits)을 확률 분포로 변환한 후, 실제 레이블과 비교하여 손실을 계산합니다.



# torch.optim.SGD:

# 옵티마이저는 모델의 파라미터(가중치)를 학습시키기 위해 사용됩니다. 여기서는 Stochastic Gradient Descent (SGD) 알고리즘이 사용되었습니다.
# SGD는 경사 하강법(Gradient Descent)의 변형으로, 전체 데이터셋 대신 배치(batch) 단위로 가중치를 업데이트합니다. 이는 학습이 빠르고, 큰 데이터셋에서도 효율적으로 작동할 수 있습니다.
# model.parameters():

# 모델의 모든 학습 가능한 파라미터(가중치와 편향)를 옵티마이저에 전달합니다. 이 파라미터들은 학습 과정에서 업데이트됩니다.
# lr=1e-3:

# 학습률(Learning Rate): 모델의 가중치를 얼마나 크게 업데이트할지를 결정하는 하이퍼파라미터입니다.
# 여기서는 학습률이 0.001로 설정되어 있습니다. 학습률이 너무 크면 학습이 불안정해지고, 너무 작으면 학습 속도가 느려질 수 있습니다.
````

- 각 학습 단계(training loop)에서 모델(배치(batch)로 제공된)은 학습 데이터 셋에 대한 예측을 수행하고, 예측 오류를 역전파하여 모델의 매개변수를 조정

````python
def train(dataloader, model, loss_fn, optimizer):           ## dataloader.dataset은 전체 데이터셋을 참조
    size=len(dataloader.dataset)                            ## 학습에 사용되는 데이터의 총 개수
    for batch, (X, y) in enumerate(dataloader):
        X, y=X.to(device), y.to(device)

        # 예측 오류 계산
        pred=model(X)
        loss=loss_fn(pred, y)

        # 역전파
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), (batch + 1) * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
````

## 각 부분의 역할
- size = len(dataloader.dataset):
  - dataloader.dataset은 전체 데이터셋을 참조합니다. size는 전체 데이터셋의 크기, 즉 학습에 사용되는 데이터의 총 개수를 나타냅니다.

- for batch, (X, y) in enumerate(dataloader)::
  - dataloader는 미니 배치(mini-batch) 단위로 데이터를 반복해서 불러옵니다. X는 입력 데이터(이미지), y는 해당 데이터의 레이블(정답)입니다.
  - enumerate(dataloader)는 배치의 인덱스(batch)와 배치 데이터(X, y)를 동시에 제공합니다.

- X, y = X.to(device), y.to(device):
  - X와 y를 사용 중인 장치(device), 즉 CPU나 GPU로 이동시킵니다. 이 작업은 데이터와 모델이 동일한 장치에 있어야 효율적으로 계산이 이루어지기 때문에 필요합니다.

- 예측 및 손실 계산:
  - pred = model(X): 모델을 통해 입력 데이터 X에 대한 예측 값을 계산합니다.
  - loss = loss_fn(pred, y): 예측 값(pred)과 실제 레이블(y) 간의 차이를 손실 함수(loss_fn)를 사용해 계산합니다. 여기서 loss_fn은 앞서 정의한 nn.CrossEntropyLoss()입니다.

- 역전파 및 옵티마이저 스텝:
  - optimizer.zero_grad(): 역전파 단계 전에 옵티마이저의 모든 경사도를 0으로 초기화합니다. 이는 이전 배치의 그래디언트가 누적되지 않도록 방지합니다.
  - loss.backward(): 손실 값을 기준으로 역전파(backpropagation)를 수행하여, 각 파라미터에 대한 그래디언트를 계산합니다.
  - optimizer.step(): 옵티마이저가 각 파라미터를 업데이트합니다. 이 단계에서 모델이 손실을 줄이는 방향으로 학습됩니다.

- 학습 상태 출력:
  - if batch % 100 == 0:: 100번째 배치마다 학습 상태를 출력합니다.
  - loss.item(): 손실 값을 Python 숫자로 변환합니다.
  - current = (batch + 1) * len(X): 현재까지 처리된 데이터 샘플 수를 계산합니다. len(X)는 현재 배치의 크기입니다.
  - print(f"loss: {loss:>7f} [{current:>5d}/{size:>5d}]"): 현재 배치의 손실 값과 학습이 진행된 데이터 수를 출력합니다. 예를 들어, loss: 0.123456 [6400/60000]과 같이 출력될 수 있습니다.


##### 전체 동작 요약
이 train 함수는 주어진 데이터 로더(dataloader)를 사용해 모델을 한 에포크(epoch) 동안 학습시킵니다. 과정은 다음과 같습니다:

- 데이터 로딩: 미니 배치 단위로 데이터를 불러옵니다.
- 예측 및 손실 계산: 모델을 통해 예측 값을 생성하고, 그 예측 값과 실제 레이블 간의 손실을 계산합니다.
- 역전파 및 파라미터 업데이트: 역전파를 통해 모델의 파라미터를 업데이트하여 손실을 줄이도록 학습합니다.
- 학습 상태 출력: 학습 과정 중에 일정 간격으로 현재 손실과 진행 상황을 출력하여 학습 상태를 모니터링할 수 있게 합니다.
이 함수는 모델 학습 과정의 핵심 부분을 담당하며, 반복적으로 호출되어 모델을 최적화하게 됩니다.

- 모델이 학습하고 있는지를 확인하기 위해 테스트 테이터셋으로 모델의 성능을 확인합니다.

````python
def test(dataloader, model, loss_fn):
    size=len(dataloader.dataset)
    num_batches=len(dataloader)
    model.eval()
    test_loss, correct=0,0
    with torch.no_grad():
        for X, y in dataloader:
            X, y=X.to(device), y.to(device)
            pred=model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
````

## 각 부분의 역할
1. size = len(dataloader.dataset):
- 테스트 데이터셋의 전체 크기를 저장합니다. size는 테스트 데이터셋에 있는 샘플의 총 개수를 나타냅니다.

2. num_batches = len(dataloader):
- num_batches는 테스트 데이터셋을 몇 개의 배치로 나누어 평가할지를 나타냅니다. 이는 데이터 로더에 설정된 배치 크기에 따라 결정됩니다.

3. model.eval():
- 모델을 평가 모드로 전환합니다. eval() 모드는 드롭아웃(dropout)이나 배치 정규화(batch normalization) 같은 레이어의 동작 방식을 변경하여 평가 시 일관된 결과를 보장합니다.

4. test_loss, correct = 0, 0:
- 테스트 손실(test_loss)과 정확히 예측한 샘플의 수(correct)를 저장할 변수를 초기화합니다.

5. with torch.no_grad()::
- 이 블록 내에서는 자동 미분(gradient)을 비활성화합니다. 이는 테스트 중에는 모델의 파라미터를 업데이트할 필요가 없기 때문에 메모리 사용량을 줄이고 계산 속도를 높이기 위해 사용됩니다.

6. for X, y in dataloader::
- 테스트 데이터 로더에서 배치 단위로 데이터를 가져옵니다. X는 입력 이미지, y는 해당 이미지의 레이블입니다.

7. X, y = X.to(device), y.to(device):
- 입력 데이터(X)와 레이블(y)를 사용 중인 장치(device), 즉 CPU나 GPU로 이동시킵니다.

8. pred = model(X):
- 모델을 사용해 입력 데이터 X에 대한 예측을 수행합니다.

9. test_loss += loss_fn(pred, y).item():
- 예측값(pred)과 실제 레이블(y) 간의 손실을 계산하고, 이를 test_loss에 누적합니다. .item()을 사용해 손실 값을 파이썬 숫자 형식으로 변환합니다.

10. correct += (pred.argmax(1) == y).type(torch.float).sum().item():
- pred.argmax(1)은 예측된 클래스 레이블을 반환합니다. 모델의 출력은 각 클래스에 대한 확률(또는 로짓)로 구성되며, argmax(1)은 가장 높은 값을 가진 클래스 인덱스를 선택합니다.
- 이 인덱스와 실제 레이블 y가 일치하는지 비교하여, 맞춘 예측의 수를 correct에 누적합니다.

11. test_loss /= num_batches:
- 전체 테스트 배치에 대한 평균 손실을 계산합니다.

12. correct /= size:
- 정확도를 계산합니다. correct는 맞춘 예측의 수, size는 전체 테스트 데이터셋의 크기이므로, 이를 통해 정확도를 구할 수 있습니다.

13. print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n"):
- 테스트 결과를 출력합니다. 정확도는 퍼센트(%)로 표시되며, 평균 손실도 출력됩니다.
- 예를 들어, "Accuracy: 85.3%, Avg loss: 0.234567"과 같이 출력될 수 있습니다.


##### 전체 동작 요약
- 모델 평가 모드 설정: 학습과는 다른 평가 모드로 모델을 전환합니다.
- 데이터 로딩 및 예측: 테스트 데이터셋에서 배치 단위로 데이터를 가져와 예측을 수행합니다.
- 손실 및 정확도 계산: 모델의 예측 결과를 기반으로 손실과 정확도를 계산합니다.
- 결과 출력: 계산된 평균 손실과 정확도를 출력하여, 모델이 테스트 데이터셋에서 얼마나 잘 작동하는지를 평가합니다.

````python
epochs=5
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_dataloader, model, loss_fn, optimizer)
    test(test_dataloader, model, loss_fn)
print("Done!")
````

1. epochs = 5:
- 모델을 학습시킬 에포크(epoch) 수를 설정합니다. 여기서는 5번의 에포크 동안 학습을 진행합니다.
- 에포크란 데이터셋 전체를 한 번 학습하는 과정을 의미합니다. 여러 에포크 동안 학습을 반복함으로써 모델이 점점 더 좋은 성능을 보이도록 합니다.

2. for t in range(epochs)::
- 5번 반복되는 학습 루프를 정의합니다. 각 반복은 하나의 에포크를 의미합니다.

3. print(f"Epoch {t+1}\n-------------------------------"):
- 현재 에포크를 출력합니다. 예를 들어, 첫 번째 에포크에서는 "Epoch 1"이 출력됩니다.
- t+1은 에포크 번호를 1부터 시작하게 합니다. range(epochs)는 0부터 시작하므로, t+1을 통해 사람에게 익숙한 1부터 시작하는 번호로 출력합니다.

4. train(train_dataloader, model, loss_fn, optimizer):
- 학습 단계: train 함수가 호출되어, 주어진 학습 데이터 로더(train_dataloader)를 사용해 모델을 한 에포크 동안 학습시킵니다.
- 이 함수는 입력 데이터(X)와 레이블(y)을 통해 모델의 예측 값을 계산하고, 그 결과를 바탕으로 모델의 파라미터(가중치)를 업데이트합니다.

5. test(test_dataloader, model, loss_fn):
- 평가 단계: test 함수가 호출되어, 테스트 데이터 로더(test_dataloader)를 사용해 학습된 모델을 평가합니다.
- 이 함수는 테스트 데이터셋에 대한 모델의 예측 정확도와 평균 손실을 계산하여 출력합니다.

6. print("Done!"):

#### 이제 이모델을 사용해서 예측을 할 수 있음

````python
classes = [
    "T-shirt/top",
    "Trouser",
    "Pullover",
    "Dress",
    "Coat",
    "Sandal",
    "Shirt",
    "Sneaker",
    "Bag",
    "Ankle boot",
]
model.eval() # 평가모드로 설정
x,y = test_data[0][0], test_data[0][1]      ## 테스트 데이터셋에서 첫번째 샘플 이미지 ([][0])과 그에 해당하는 레이블([][1])을 가져옴
with torch.no_grad():                       ## 평가 또는 추론 시에는 모델을 업데이트할 필요가 없으므로 그라디엔트를 계산할 필요 없음
    x = x.to(device)                        ## 테스트 샘플을 현재 설정된 장치(CPU 또는 GPU)로 이동
    pred = model(x)
    predicted, actual = classes[pred[0].argmax(0)], classes[y]    ## 모델의 예측 결과 중 가장 높은 값을 가지는 인덱스와 실제 레이블에 해당하는 클래스 이름
    print(f'Predicted: "{predicted}", Actual: "{actual}"')
````

- x, y = test_data[0][0], test_data[0][1]:
  - test_data[0]는 테스트 데이터셋에서 첫 번째 샘플을 가져옵니다.
  - x는 이미지 데이터(입력), y는 해당 이미지의 실제 레이블입니다.
  - test_data[0][0]은 첫 번째 샘플의 이미지(텐서)를 의미하고, test_data[0][1]은 첫 번째 샘플의 레이블을 의미합니다.

- with torch.no_grad()::
  - 이 블록 내에서는 자동 미분(gradient)을 비활성화합니다. 이는 모델을 평가하거나 추론할 때 불필요한 계산을 방지하고 메모리 사용을 줄이기 위해 사용됩니다.

- x = x.to(device):
  - 입력 이미지 x를 현재 설정된 장치(device), 즉 CPU 또는 GPU로 이동시킵니다. 모델이 이 장치에서 실행되기 때문에, 입력 데이터도 같은 장치로 이동시켜야 합니다.

- pred = model(x):
  - 모델에 입력 이미지 x를 전달하여 예측 값을 계산합니다.
  - pred는 모델의 출력으로, 각 클래스에 대한 로짓(logits)을 포함하는 텐서입니다. 이 로짓은 아직 확률로 변환되지 않은 상태

- pred[0].argmax(0):
  - pred[0]는 모델의 출력 텐서에서 첫 번째(이 경우 유일한) 예측을 의미합니다.
  - argmax(0)은 가장 높은 값을 가진 인덱스를 반환합니다. 이는 모델이 예측한 클래스 레이블입니다.

- classes[pred[0].argmax(0)]:
  - 모델이 예측한 클래스 인덱스를 사용하여, 해당 클래스 이름을 classes 리스트에서 가져옵니다. 이것이 predicted입니다.

- classes[y]:
  - 실제 레이블 y에 해당하는 클래스 이름을 classes 리스트에서 가져옵니다. 이것이 actual입니다.

- torch: 이것은 메인 PyTorch 라이브러리입니다. 여기에는 GPU를 통한 가속을 통한 텐서 계산 지원, 신경망 훈련을 용이하게 하는 자동 차별화, 모델 구축 및 훈련을 위한 다양한 유틸리티가 포함됩니다.
- torch.nn: 레이어, 활성화 함수, 손실 함수와 같은 신경망의 구성 요소를 제공하는 PyTorch의 하위 모듈입니다. 신경망의 아키텍처를 정의하는 데 필수적입니다.
- torch.nn.function: 이 모듈에는 torch.nn 레이어에서 사용되는 기능이 포함되어 있습니다. 입력 데이터 및 가중치에 이러한 함수를 직접 사용할 수 있으므로 일부 작업에 더 많은 유연성을 제공합니다. 여기에는 활성화, 손실 계산 및 상태(즉, 가중치)를 유지하지 않는 다양한 기타 작업을 위한 함수가 포함됩니다.
- torch.optim: 이 하위 모듈은 SGD(Stochastic Gradient Descent), Adam 등과 같은 신경망 훈련을 위한 최적화 알고리즘을 제공합니다. 이러한 최적화 프로그램은 계산된 기울기를 기반으로 네트워크의 가중치를 업데이트하는 데 사용됩니다.
- torchvision: 이미지 데이터 작업을 위한 유틸리티를 제공하는 PyTorch 프로젝트의 패키지입니다. 여기에는 사전 정의된 데이터세트(예: MNIST, CIFAR10, FashionMNIST), 모델 아키텍처(예: ResNet, AlexNet) 및 전처리를 위한 일반적인 이미지 변환이 포함됩니다.
- torchvision.transforms: 일반적인 이미지 변환을 제공하는 torchvision 내의 모듈입니다. 이는 이미지를 신경망에 공급하기 전에 데이터 증대 및 이미지 전처리에 사용될 수 있습니다. 예로는 크기 조정, 정규화, 텐서로 변환 등이 있습니다.
- SubsetRandomSampler: 교체 없이 데이터세트에서 요소를 무작위로 샘플링하는 데 사용되는 도구입니다. 데이터 세트를 훈련 및 검증/테스트 세트로 분할하거나 모델 훈련을 위해 사용자 정의 데이터 샘플링 전략을 구현하려는 경우에 특히 유용

######### One Cell로 통합

````python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor

## 공개 데이터셋에서 학습 데이터를 down
training_data=datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,        ## 데이터셋이 로컬에 없으면 자동으로 인터넷에서 다운로드
    transform=ToTensor(), ## 이미지 데이터를 PyTorch 텐서로 변환; 0255의 픽셀 값--> 0/1 사이의 값으로 정규화하고 텐서 형태로 변환, 모델에 입력할 수 있게 변환
)

## 공개 데이터셋에서 테스트 데이터를 down
test_data=datasets.FashionMNIST(
    root="data",
    train=False,        ## 테스트 데이터
    download=True,      ## 이미 다운로드된 경우, 해당 디렉토리에서 데이터를 불러옵
    transform=ToTensor(),
)

batch_size=64

# 데이터 로더를 생성
train_dataloader=DataLoader(training_data, batch_size=batch_size)
test_dataloader=DataLoader(test_data, batch_size=batch_size)

for X, y in test_dataloader:
    print("Shape of X [N, C, H, W]: ", X.shape)
    print("Shape of y: ", y.shape, y.dtype)
    break

# 학습에 사용할 CUP나 MPS 장치를 설정
device = (
    "cuda"
    if torch.cuda.is_available()
    # else "mps"
    # if torch.backends.mps.is_available()            ## mps는 Apple의 M1/M2 칩에서 제공하는 GPU 가속 기능
    else "cpu"
)
print(f"Using {device} device")

# 모델을 정의
class NeuralNetwork(nn.Module):     # NeuralNetwork 클래스: PyTorch의 신경망 모듈(nn.Module)을 상속하여 간단한 신경망을 정의
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()         ## 28x28 크기의 2D 이미지를 1D 벡터로 변환합니다. 예를 들어, [28, 28] 크기의 이미지를 [784] 크기의 벡터로 변환
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),          ## 입력 노드가 784(28x28)개이고 출력 노드가 512개인 선형 계층(fully connected layer)
            nn.ReLU(),
            nn.Linear(512, 512),            ## 두 번째 선형 계층으로, 512개의 입력을 받아 512개의 출력을 생성
            nn.ReLU(),
            nn.Linear(512, 10)              # 마지막 선형 계층, 512개의 입력을 받아 10개의 출력을 생성. 이 10개의 출력은 이미지가 10개 클래스 중 어떤 클래스에 속하는지를 나타냄
        )

    def forward(self, x):
        x = self.flatten(x)                 # 입력 이미지를 평평하게 만들어 1D 벡터로 변환
        logits = self.linear_relu_stack(x)  # 변환된 입력을 위에서 정의한 계층에 통과시켜 결과값(logits)을 얻습
        return logits

model = NeuralNetwork().to(device)          ## NeuralNetwork 클래스의 인스턴스를 생성한 후, 지정된 장치(device)로 모델을 이동
print(model)

loss_fn=nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(model.parameters(), lr=1e-3)

def train(dataloader, model, loss_fn, optimizer):           ## dataloader.dataset은 전체 데이터셋을 참조
    size=len(dataloader.dataset)                            ## 학습에 사용되는 데이터의 총 개수
    for batch, (X, y) in enumerate(dataloader):
        X, y=X.to(device), y.to(device)

        # 예측 오류 계산
        pred=model(X)
        loss=loss_fn(pred, y)

        # 역전파
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), (batch + 1) * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")


def test(dataloader, model, loss_fn):
    size=len(dataloader.dataset)
    num_batches=len(dataloader)
    model.eval()
    test_loss, correct=0,0
    with torch.no_grad():
        for X, y in dataloader:
            X, y=X.to(device), y.to(device)
            pred=model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")



epochs=5
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_dataloader, model, loss_fn, optimizer)
    test(test_dataloader, model, loss_fn)
print("Done!")


####### 모델 저장하기
import os
# defice the save path
save_path="/content/drive/MyDrive/KITA_2024/M6_DeepLearning/DATA/model/model.pth"

# ensure dir exists
os.makedirs(os.path.dirname(save_path), exist_ok=True)

torch.save(model.state_dict(), save_path)   ## model.state_dic() 모델의 가중치만 저장함
print("Saved PyTorch Model State to model.pth")

####### 모델 불러오기
model=NeuralNetwork().to(device)     ## 모델의 가중치만 저장되어 있으니 불러올때는 모델을 만들어놓고 가중치를 업데이트 하는 방식
model.load_state_dict(torch.load(save_path))

####### 모델을 사용해서 예측
classes = [
    "T-shirt/top",
    "Trouser",
    "Pullover",
    "Dress",
    "Coat",
    "Sandal",
    "Shirt",
    "Sneaker",
    "Bag",
    "Ankle boot",
]
model.eval() # 평가모드로 설정
x,y = test_data[0][0], test_data[0][1]      ## 테스트 데이터셋에서 첫번째 샘플 이미지 ([][0])과 그에 해당하는 레이블을 가져옴
with torch.no_grad():                       ## 평가 또는 추론 시에는 모델을 업데이트할 필요가 없으므로 그라디엔트를 계산할 필요 없음
    x = x.to(device)                        ## 테스트 샘플을 현재 설정된 장치(CPU 또는 GPU)로 이동
    pred = model(x)
    predicted, actual = classes[pred[0].argmax(0)], classes[y]    ## 모델의 예측 결과 중 가장 높은 값을 가지는 인덱스와 실제 레이블에 해당하는 클래스 이름
    print(f'Predicted: "{predicted}", Actual: "{actual}"')
````
