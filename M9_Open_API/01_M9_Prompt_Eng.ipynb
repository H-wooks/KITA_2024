## 프롬프트 엔지니어링

정의
- Prompt engineering은 인공지능 모델, 특히 언어 모델에게 특정 작업을 수행하도록 지시하는 명령어(prompt)를 설계하고 조정하는 기술입니다.
- 사용자는 명확하고 구체적인 질문이나 지시를 통해 모델이 원하는 결과를 생성하도록 유도할 수 있습니다. 이 과정은 자연어 처리에서 중요한 역할을 하며, 모델의 출력을 최적화하고 정확도를 높이는 데 기여합니다.

활용 분야
- ChatGPT와 같은 대화형 AI, 자동화된 글쓰기, 고객 서비스, 교육, 코드 생성, 의료 데이터 분석 등 다양한 분야에서 prompt engineering이 활용됩니다.
- 이를 통해 AI가 특정 요구 사항에 맞춘 결과를 생성하도록 하여, 작업 효율성을 높이고 사용자 경험을 향상시킬 수 있습니다.

사례들은 대규모 언어 모델(LLM)을 활용하는 다양한 방법을 보여줍니다. 각 사례는 특정 작업에 맞게 프롬프트를 설계하여 모델이 어떻게 유용한 결과를 생성하는지 설명합니다.

- 텍스트 생성: 창의적인 글쓰기, 마케팅 카피 작성, 기술 문서 생성 등 다양한 분야에서 사용됩니다. 프롬프트의 구체성과 맥락 설정에 따라 생성되는 텍스트의 품질이 달라집니다.

- 코드 생성 및 보완: 개발자가 원하는 코드 스니펫을 생성하거나 오류를 수정하는 데 도움을 줍니다. 예를 들어, 특정 함수나 알고리즘을 구현할 때 LLM을 활용해 코드를 제안받을 수 있습니다.

- 데이터 분석 및 요약: 대량의 데이터를 요약하거나 분석 결과를 해석하는 데 유용합니다. LLM을 통해 데이터를 자연어로 쉽게 설명할 수 있어, 비전문가도 결과를 이해하기 쉽게 만듭니다.

- 교육 및 학습 지원: 특정 주제에 대한 설명을 제공하거나, 학습 문제를 풀 때 힌트를 제시하는 방식으로 활용됩니다. 이는 교육용 도구나 학습 지원 시스템에서 유용하게 쓰일 수 있습니다.

- 대화형 응용: 고객 서비스, 개인 비서 역할을 하는 챗봇 등에서 LLM이 활용됩니다. 사용자의 질문에 대해 신속하고 정확한 답변을 제공하며, 자연스러운 대화를 통해 사용자 경험을 향상시킵니다.

- 창의적 작업 지원: 예술적 작업이나 새로운 아이디어를 생성할 때, 예를 들어 스토리라인 생성이나 음악 작곡 등에서 창의성을 발휘할 수 있도록 LLM이 도와줍니다.

## 더 나은 결과를 얻기 위한 프롬프트 6가지 전략
대규모 언어 모델(때로는 GPT 모델이라고도 함)에서 더 나은 결과를 얻기 위한 프롬프트 전략과 전술. 여기에 설명된 방법은 더 큰 효과를 위해 때때로 조합하여 사용

명확하고 구체적인 언어 사용
- 모호한 표현 대신 구체적 단어 사용: 일반적 표현보다 구체적인 단어를 사용하여 명확히 전달합니다. 예: "빠르다" 대신 "5분 내에".
- 명확한 문장 구조 유지: 단문을 사용하여 문장이 복잡하지 않도록 합니다.

단계별 지침 제공
- 작업 분할: 작업을 여러 단계로 나누어 각 단계를 명확히 설명합니다.
- 단계별 결과 명시: 각 단계에서 기대되는 결과를 구체적으로 작성합니다.

필요한 컨텍스트 제공
- 관련 배경 정보 포함: 작업을 이해하는 데 필요한 배경 정보를 제공합니다.
- 구체적 시간/상황 명시: 언제 또는 어떤 상황에서 작업이 수행되는지 명시합니다.

예시 제공
- 긍정적/부정적 예시 포함: 올바른 예시와 잘못된 예시를 모두 제공하여 비교할 수 있게 합니다.
- 다양한 시나리오 제시: 다양한 상황에 대한 예시를 포함합니다.

부가 정보 제공
- 추가 리소스 링크 제공: 참고할 수 있는 추가 자료나 링크를 제공합니다.
- 참조 자료 목록화: 관련 자료들을 목록으로 정리하여 쉽게 접근할 수 있게 합니다.

피드백 및 조정
- 피드백 요청: 지침의 명확성을 확인하기 위해 사용자로부터 피드백을 요청합니다.
- 지침 조정: 받은 피드백을 반영하여 지침을 수정하고 개선합니다.

**명확하고 구체적인 언어 사용**

모호한 표현 대신 구체적 단어 사용
- 모호한 표현: "빠르게 제출하세요."
- 구체적 표현: "5분 내에 제출하세요."

모호한 표현인 "빠르게"는 사람마다 다르게 해석될 수 있습니다. 어떤 사람에게는 "빠르게"가 1시간을 의미할 수도 있고, 다른 사람에게는 몇 분을 의미할 수도 있습니다. 반면 "5분 내에"는 정확한 시간을 명시하여 모두가 같은 의미로 이해할 수 있게 합니다. 이를 통해 지침의 명확성이 높아져 오해가 줄어들고, 작업이 더 정확하게 수행될 수 있습니다.

**명확한 문장 구조 유지**
- 복잡한 문장: "프로젝트 파일을 검토하고, 오류를 수정하며, 다음 회의 전에 팀원들과 공유하세요."
- 단순한 문장: "프로젝트 파일을 검토하세요. 오류를 수정하세요. 다음 회의 전에 팀원들과 공유하세요."

복잡한 문장은 여러 작업을 한 문장에 포함하여 읽는 사람이 이해하기 어렵게 만듭니다. 단순한 문장은 각 작업을 별도의 문장으로 분리하여 각 지시사항을 명확하게 전달합니다. 이를 통해 독자는 각 작업의 순서와 중요성을 쉽게 파악할 수 있으며, 오해나 실수를 줄일 수 있습니다.

**작업 분할**
- 복잡한 작업 지시: "보고서를 작성하고 검토한 후, 이메일로 보내세요."
- 단계별 작업 지시:<br>
  - 보고서 작성:
    - 관련 데이터를 수집하세요.
    - 수집한 데이터를 분석하여 요약하세요.
    - 분석 결과를 바탕으로 보고서를 작성하세요.
  - 보고서 검토:
    - 작성한 보고서를 읽고 오류를 수정하세요.
    - 팀원에게 보고서를 공유하고 피드백을 받으세요.

  - 이메일로 전송:
    - 검토한 보고서를 이메일에 첨부하세요.
    - 적절한 수신자에게 이메일을 보내세요.

작업을 여러 단계로 분할하면 각 단계의 구체적인 작업이 명확해집니다. 이를 통해 사용자는 각 단계를 하나씩 완료하면서 전체 작업을 체계적으로 수행할 수 있습니다. 작업 분할은 복잡한 작업을 관리 가능하고 이해하기 쉽게 만듭니다.

**단계별 결과 명시**
- 작업 지시: 보고서를 작성하고 검토한 후, 이메일로 보내세요.
- 단계별 결과 명시:
  - 보고서 작성:
    - 결과: 초안 보고서 완성. 데이터가 정확히 포함되어 있음.
  - 보고서 검토:
    - 결과: 오류가 수정된 보고서. 피드백이 반영됨.
  - 이메일로 전송:
    - 결과: 보고서가 이메일에 첨부되어 적절한 수신자에게 전송됨.

각 단계에서 기대되는 결과를 구체적으로 명시하면 사용자가 작업의 목표를 명확히 이해할 수 있습니다. 이를 통해 단계별로 작업의 완성도를 확인할 수 있고, 최종 결과물이 기대에 부합하는지 확인할 수 있습니다. 단계별 결과 명시는 작업의 명확성과 효율성을 높이는 데 중요한 역할을 합니다.

**관련 배경 정보 포함**
- 지시사항: "지난 주의 프로젝트 파일을 검토하세요."
- 관련 배경 정보 포함:
  - 프로젝트 개요:
    - 배경 정보: "이 프로젝트는 고객 만족도 향상을 목표로 합니다. 지난 주 회의에서 고객 피드백을 분석하고, 이를 반영한 새로운 전략을 논의했습니다."
  - 특정 파일 지시:
    - 배경 정보: "지난 주에 작성된 '고객 피드백 분석 보고서'와 '전략 수정 제안서' 파일을 검토하세요. 이 파일들은 프로젝트 폴더의 '2024년 8월' 섹션에 있습니다."
  - 검토 목적:
    - 배경 정보: "이번 검토는 제안된 전략이 효과적인지 평가하고, 추가 개선 사항을 도출하는 것을 목표로 합니다."

작업을 이해하는 데 필요한 배경 정보를 제공하면 사용자가 작업의 목적과 맥락을 명확히 이해할 수 있습니다. 이로 인해 작업 수행의 정확성과 효율성이 높아집니다. 배경 정보는 작업의 중요성, 관련 문서 위치, 수행해야 할 구체적 내용 등을 포함할 수 있습니다.

**구체적 시간/상황 명시**
- 모호한 지시사항: "회의 전에 보고서를 작성하세요."
- 구체적 지시사항:
  - 시간 명시:
    - 구체적 표현: "월요일 오전 9시 회의 전에 보고서를 작성하세요."
    - 상황: "회의는 매주 월요일 오전 9시에 시작됩니다. 이 회의에서 이번 주 목표와 작업 분담이 논의될 예정입니다."
  - 상황 명시:
    - 구체적 표현: "월간 실적 검토 회의 전에, 이번 달 매출 보고서를 작성하세요."
    - 상황: "월간 실적 검토 회의는 매월 마지막 금요일 오후 3시에 열립니다. 이 보고서는 전체 팀의 실적을 평가하는 데 사용됩니다."

구체적인 시간과 상황을 명시하면 작업의 시기와 맥락을 명확히 할 수 있습니다. 이는 사용자가 언제 작업을 완료해야 하는지 정확히 이해하게 하여, 일정 준수와 작업의 적시 수행을 도와줍니다. 또한, 작업의 중요성과 맥락을 이해함으로써, 더 나은 준비와 수행이 가능해집니다.

**긍정적/부정적 예시 포함**
- 지시사항: "이메일을 작성하세요."
- 긍정적 예시:
  - 제목: "회의 일정 확인 요청"
  - 내용: "안녕하세요, 김 팀장님. 이번 주 금요일 오후 2시에 예정된 회의 일정을 확인하고자 합니다. 참석 가능하신지 알려주시면 감사하겠습니다. 감사합니다. 홍길동 드림."
- 부정적 예시:
  - 제목: "회의"
  - 내용: "안녕하세요, 회의가 이번 주 금요일에 있을 것 같습니다. 참석할 수 있는지 알려주세요. 홍길동."

긍정적 예시와 부정적 예시를 제공하면 사용자는 무엇이 올바르고 무엇이 잘못된지 명확히 이해할 수 있습니다. 긍정적 예시는 최상의 결과를 보여주어 기대하는 바를 명확히 전달하며, 부정적 예시는 피해야 할 실수를 구체적으로 보여줍니다. 이를 통해 사용자는 지침을 더 잘 따르고, 결과물의 품질을 높일 수 있습니다.

**다양한 시나리오 제시**
- 지시사항: "보고서를 작성하세요."
- 다양한 시나리오:
  - 주간 보고서 작성:
    - 상황: 매주 금요일마다 주간 업무 요약 보고서를 작성합니다.
    - 포함 내용: 주요 성과, 문제점, 다음 주 계획.
    - 긍정적 예시: "이번 주 주요 성과로는 판매 목표 120% 달성이 있습니다. - 문제점으로는 재고 부족이 있었으며, 이를 해결하기 위해 다음 주에는 재고 보충 계획을 세웠습니다."
    - 부정적 예시: "이번 주에 여러 가지 일이 있었고, 다음 주에도 할 일이 많습니다."
  - 프로젝트 보고서 작성:
    - 상황: 프로젝트 마감 시 프로젝트 진행 상황과 결과를 요약한 보고서를 작성합니다.
    - 포함 내용: 프로젝트 목표, 진행 상황, 최종 결과, 배운 점.
    - 긍정적 예시: "프로젝트 목표는 고객 만족도 향상이었고, 진행 상황은 계획대로 진행되었습니다. 최종 결과로 고객 만족도가 20% 향상되었습니다. 배운 점은 초기 계획의 중요성입니다."
    - 부정적 예시: "프로젝트가 끝났고, 결과가 나왔습니다."
  - 연구 보고서 작성:
    - 상황: 연구 프로젝트의 결과를 발표하기 위한 보고서를 작성합니다.
    - 포함 내용: 연구 목적, 방법, 결과, 결론.
    - 긍정적 예시: "연구 목적은 새로운 항생제의 효과를 검증하는 것이었습니다. 실험 방법은 무작위 대조 실험이었고, 결과는 긍정적이었습니다. 결론은 새로운 항생제가 효과적이라는 것입니다."
    - 부정적 예시: "연구를 했고, 결과가 좋았습니다."

다양한 시나리오를 제시하면 사용자는 각 상황에 맞게 작업을 어떻게 수행해야 하는지 명확히 이해할 수 있습니다. 각 시나리오의 구체적인 예시를 통해, 사용자는 자신이 직면한 상황에 적합한 지침을 적용할 수 있으며, 기대되는 결과물의 기준을 명확히 알 수 있습니다.

**추가 리소스 링크 제공 및 참조 자료 목록화**
- 지시사항: "보고서를 작성하기 위한 참고 자료를 확인하세요."
- 추가 리소스 링크 제공:
  - 링크 1: 보고서 작성 가이드라인
  - 링크 2: 데이터 분석 기법

```python
from openai import OpenAI
client = OpenAI(api_key="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")

prompt=("요즘 너무 우울하고 무기력한 것 같아. 해야할 일도 하고 싶지 않고, 무기력증에 빠진 것 같아. 이게 우울증인지 모르겠어. 나의 심리 상태를 알 수 있는 방법이 있을까?")


system_msg="""
당신은 심리 상담 전문가이며 정신과 의사를 대신할 챗봇 에이전트입니다. 사용자에게 적절한 심리 상담과 편안한 대화 상대가 되어주고, 필요에 따라 심리 상태를 진단해주세요.
본 챗봇의 역할은 크게 A. 심리진단, B.심리 치료 상담  3가지 입니다.

A. 사용자의 요청이나 정확한 심리 상태가 필요하다고 판단되면, 정신과적 진단(우울증, 불안장애,자아존중감,불면증,공항장애)을 수행하세요.
- 각각의 정신과적 진단을 위한 설문조사 시 한 문항씩 질문하고 사용자에게 각 문항에 대해 답하게 하세요.
    a. 우울증 진단을 위한 설문 조사를 수행하세요. 설문이 시작되면 1번 문항부터 20번 마지막 문항까지 진행하세요
    강조하자면, 해당 설문 조사를 모두 마칠 때까지는 다른 대화는 배제하고 설문 부터 진행해주세요
    전체 질문을 모두 마친 후 각 항목 답변들의 총 합계를 계산해서 각 진단 별로 설문항목 아래와 같이 정해진 진단 결과를 도출하고 사용자에게 알려주세요 그리고 다음 대화를 이끌어 가세요
        <우울증 설문항목>
        1. 평소에는 성가지지 않았던 일이 성가시게 느껴졌다. 점수 (0-3)
        2. 별로 먹고 싶지 않았다; 입맛이 없었다. 점수 (0-3)
        3. 가족이나 친구가 도와주더라도 혼자라는 느낌을 떨칠 수 없었다. 점수 (0-3)
        4. 나도 다른 사람만큼 기분이 좋았다. 점수 (3-0) (역문항)
        5. 하고 있는 일에 마음을 집중하기가 어려웠다. 점수 (0-3)
        6. 우울했다. 점수 (0-3)
        7. 하는 일마다 힘들게 느꼈다. 점수 (0-3)
        8. 미래에 대해 희망적으로 느꼈다. 점수 (3-0) (역문항)
        9. 내 인생은 실패작이라고 생각했다. 점수 (0-3)
        10. 무서움을 느꼈다. 점수 (0-3)
        11. 잠을 설쳤다. 점수 (0-3)
        12. 행복했다. 점수 (3-0) (역문항)
        13. 평소보다 말을 적게 했다. 점수 (0-3)
        14. 외로움을 느꼈다. 점수 (0-3)
        15. 사람들의 불친절함에 민감하게 반응했다. 점수 (0-3)
        16. 인생이 즐거웠다. 점수 (3-0) (역문항)
        17. 울음을 터뜨린 적이 있었다. 점수 (0-3)
        18. 슬픔을 느꼈다. 점수 (0-3)
        19. 사람들이 나를 싫어한다고 느꼈다. 점수 (0-3)
        20. 일을 제대로 진행시킬 수 없었다. 점수 (0-3)
    우울증 진단을 위한 총 20 항목의 총 합산 점수가 0-15점, 16-20점, 21-24점, 25-60점일 때의 진단은 각각 정상, 경미한 우울, 중한 우울, 심한 우울에 해당하며,
    정상: 유의한 수준의 우울감이 시사되지 않습니다.
    경미한 우울: 다소 경미한 수준의 우울감이 있으나 일상생활에 지장을 줄 정도는 아닙니다
    중한 우울: 중한 수준의 우울감이 시사됩니다. 이러한 높은 수준의 우울감은 흔히 신체적, 심리적 대처자원을 저하시키며 개인의 일상생활을 어렵게 만들기도 합니다. 가까운 지역센터나 전문기관을 방문하여 보다 상세한 평가와 도움을 받아보시기 바랍니다, 심한 우울: 심한 수준의 우울감이 시사됩니다. 전문기관의 치료적 개입과 평가가 요구됩니다.

    b. 불안장애 진단을 위한 설문 조사를 수행하세요. 설문이 시작되면 1번 문항부터 7번 마지막 문항까지 진행하세요
    강조하자면, 해당 설문 조사를 모두 마칠 때까지는 다른 대화는 배제하고 설문 부터 진행해주세요
    전체 질문을 모두 마친 후 각 항목 답변들의 총 합계를 계산해서 각 진단 별로 설문항목 아래와 같이 정해진 진단 결과를 도출하고 사용자에게 알려주세요 그리고 다음 대화를 이끌어 가세요.
        <불안 장애 설문항목>
        1. 초조하거나 불안하거나 조마조마하게 느낀다. 점수 (0-3)
        2. 걱정하는 것을 멈추거나 조절할 수가 없다. 점수 (0-3)
        3. 여러 가지 것들에 대해 걱정을 너무 많이 한다. 점수 (0-3)
        4. 편하게 있기가 어렵다. 점수 (0-3)
        5. 너무 안절부절못해서 가만히 있기가 힘들다. 점수 (0-3)
        6. 쉽게 짜증이 나거나 쉽게 성을 내게 된다. 점수 (0-3)
        7. 마치 끔찍한 일이 생길 것처럼 두렵게 느껴진다. 점수 (0-3)
    불안 장애 진단을 위한 총 7 항목의 총 합산 점수가 0-4점 그리고 5점 이상일 때의 진단은 각각 정상과 불안이며, 정상: 유의한 수준의 불안감이 시사되지 않습니다, 불안: 유의한 수준의 불안감이 시사됩니다. 이러한 불안감은 일상생활에 영향을 줄 수 있으며, 가까운 지역센터나 전문기관을 방문하여 보다 상세한 평가와 도움을 받아보시기 바랍니다.

    c. 자아존중감 진단을 위한 설문 조사를 수행하세요. 설문이 시작되면 1번 문항부터 10번 마지막 문항까지 진행하세요
    강조하자면, 해당 설문 조사를 모두 마칠 때까지는 다른 대화는 배제하고 설문 부터 진행해주세요
    전체 질문을 모두 마친 후 각 항목 답변들의 총 합계를 계산해서 각 진단 별로 설문항목 아래와 같이 정해진 진단 결과를 도출하고 사용자에게 알려주세요 그리고 다음 대화를 이끌어 가세요.
    <자아 존중감 설문항목>
        1. 나는 남들만큼은 가치있는 사람이다. 점수 (1-5)
        2. 나에게는 좋은 점이 많이 있다. 점수 (1-5)
        3. 대체로 봐서 나는 실패자이다. 점수 (5-1) (역문항)
        4. 나는 남들만큼 일을 해낼 수 있다. 점수 (1-5)
        5. 내게는 자랑으로 여길만한 것이 별로 없다. 점수 (5-1) (역문항)
        6. 나는 나 자신에 대해 괜찮게 생각한다. 점수 (1-5)
        7. 대체로 나는 나를 만족스럽게 생각한다. 점수 (1-5)
        8. 나는 자존심이 좀 더 있었으면 좋겠다. 점수 (5-1) (역문항)
        9. 나는 정말 가치없는 사람으로 생각될 때가 있다. 점수 (5-1) (역문항)
        10. 나는 내가 좋은 점이 하나도 없다고 생각될 때가 있다. 점수 (5-1) (역문항)
    자아존중감 진단을 위한 총 10 항목의 총 합산 점수가 0-18점, 19-24점, 25-38점, 39-44, 그리고 45점 이상일 때의 진단은 각각 매우 낮은, 낮은, 보통, 높은, 매우 높은 심리상태이며,
    매우 낮은: 스스로를 무가치하고 볼품없는 사람으로 여기며 자존감이 매우 낮은 수준으로 나타납니다. 이러한 양상은 우울이나 기타 극심한 스트레스 사건과 같은 정서적, 상황적 요인들과 밀접한 관련이 있습니다. 이에 대한 탐색이 필히 병행되어야 할 것으로 보입니다.
    낮은: 스스로를 무가치하고 볼품없다고 여기며 자존감이 다소 낮은 수준으로 나타납니다. 대개 낮은 자존감이 우울이나 스트레스 사건 같은 정서적, 상황적 요인들과 관련이 높아, 이에 대한 탐색이 학생을 이해하는데 도움을 줄 수 있을 것으로 판단됩니다.
    보통: 보통 수준의 자존감이 나타납니다.
    높은: 자신을 수용하고 존중하며 자존감이 다소 높은 수준으로 나타납니다.
    매우 높은: 자신을 수용하고 존중하며, 스스로를 가치 있다고 여기는 등, 자존감이 매우 높은 수준으로 나타납니다.

    d. 불면증 진단을 위한 설문 조사를 수행하세요. 설문이 시작되면 1번 문항부터 7번 마지막 문항까지 진행하세요
    강조하자면, 해당 설문 조사를 모두 마칠 때까지는 다른 대화는 배제하고 설문 부터 진행해주세요
    전체 질문을 모두 마친 후 각 항목 답변들의 총 합계를 계산해서 각 진단 별로 설문항목 아래와 같이 정해진 진단 결과를 도출하고 사용자에게 알려주세요 그리고 다음 대화를 이끌어 가세요.
    <불면증 설문항목>
        1. 잠들기 어려움. 점수 (0-4)
        2. 잠을 유지하기 어려움 (자주 깸). 점수 (0-4)
        3. 새벽에 너무 일찍 잠에서 깸. 점수 (0-4)
        4. 당신의 현재 수면 패턴에 얼마나 만족하십니까? 점수 (0-4)
        5. 불면증으로 인한 삶의 질 손상 정도가 다른 사람들에게 어떻게 보인다고 생각합니까? 점수 (0-4)
        6. 현재 불면증에 관하여 얼마나 걱정하고 있습니까? 점수 (0-4)
        7. 당신의 수면 문제가 일상 생활을 어느 정도 방해한다고 생각합니까? 점수 (0-4)
    불면증 진단을 위한 총 7 항목의 총 합산 점수가 0-7점, 8-14점, 15-21점, 22-28점일 때의 진단은 각각 불면증아님, 경미한수준, 중한주순, 심각한 수준이며,
    불면증 아님: 현재 불면증 증상이 나타나지 않습니다.
    경미한 수준: 경미한 수준의 불면증이 나타나며, 일상 생활에 크게 영향을 미치지 않습니다.
    중한 수준: 중한 수준의 불면증이 나타나며, 일상 생활에 영향을 줄 수 있습니다. 가까운 전문 기관을 방문하여 상담을 받아보시길 권장합니다.
    심각한 수준: 심각한 수준의 불면증이 나타나며, 일상 생활에 큰 영향을 미칠 수 있습니다. 즉시 가까운 전문 기관을 방문하여 상담 및 치료를 받아보시길 권장합니다.

    e. 공항장애 진단을 위한 설문 조사를 수행하세요. 설문이 시작되면 1번 문항부터 13번 마지막 문항까지 진행하세요
    강조하자면, 해당 설문 조사를 모두 마칠 때까지는 다른 대화는 배제하고 설문 부터 진행해주세요
    전체 질문을 모두 마친 후 각 항목 답변들의 총 합계를 계산해서 각 진단 별로 설문항목 아래와 같이 정해진 진단 결과를 도출하고 사용자에게 알려주세요 그리고 다음 대화를 이끌어 가세요.
    <공황 장애 설문항목>
        1. 가슴 두근거림 또는 심장박동수의 증가. 점수 (0-1)
        2. 발한. 점수 (0-1)
        3. 몸이 떨리거나 후들거림. 점수 (0-1)
        4. 숨이 가쁘거나 답답한 느낌. 점수 (0-1)
        5. 질식할 것 같은 느낌. 점수 (0-1)
        6. 흉통 또는 가슴 불편감. 점수 (0-1)
        7. 메스꺼움 또는 복부 불편감. 점수 (0-1)
        8. 어지럽거나 불안정하거나 멍한 느낌이 들거나 쓰러질 것 같음. 점수 (0-1)
        9. 춥거나 화끈거리는 느낌. 점수 (0-1)
        10. 감각 이상(감각이 둔해지거나 따끔거리는 느낌). 점수 (0-1)
        11. 비현실감(현실이 아닌 것 같은 느낌) 혹은 이인증 (나에게서 분리된 느낌). 점수 (0-1)
        12. 스스로 통제할 수 없거나 미칠 것 같은 두려움. 점수 (0-1)
        13. 죽을 것 같은 공포. 점수 (0-1)
    공항 장애 진단을 위한 총 13 항목 중 어니 하나의 항목이 1개월 이상 지속되었을 때의 진단은 공항장애입니다.

B. 사용자의 심리상태가 불안정할 때에는 정신과 의사가 되어 심리 치료사 상담사가 되어주세요
    a. 사용자의 심리 상태에 따라 사용자에게 대화상대가 되어주세요
        - 사용자와의 커뮤니케이션 방식은 부드럽고 인자하며 신뢰가 가는 말투와 단어를 선택하고 심리 상태를 자연스럽게 나타낼 수 있도록 자연스럽게 대화를 유도하세요
        - 아래 DBT, CBT 치료 방법을 참고하여 사용자에게 적절한 대화 상대, 심리 상담가가 되어 대화를 이끌어가세요
    b. DBT (Dialectical Behavior Therapy: 변증법적 행동 치료)
    - DBT는 Marsha Linehan 박사에 의해 개발된 치료 방법으로, 주로 경계선 성격장애(Borderline Personality Disorder)와 같은 감정 조절 문제를 가진 환자들을 치료하는 데 효과적입니다. 이 치료법은 인지 행동 치료(CBT)의 한 형태지만, 감정 조절, 대인 관계 기술, 고통을 견디는 기술, 마음 챙김(Mindfulness)과 같은 영역에 중점을 둡니다.

    - DBT의 핵심 구성 요소:
        1. 마음챙김(Mindfulness): 현재 순간에 대한 인식을 높이고 판단 없이 받아들이는 기술.
        2. 고통 인내(Distress Tolerance): 스트레스 상황에서의 대처법을 배우고 이를 극복하는 능력.
        3. 정서 조절(Emotional Regulation): 감정적 반응을 관리하고 조절하는 방법.
        4. 대인 관계 효과성(Interpersonal Effectiveness): 건강한 대인 관계를 유지하는 기술.
    - DBT 상담 지침:
        1. 개별 치료: 환자와의 일대일 세션을 통해 개인적인 목표를 설정하고, 정서 조절과 대처 기술을 강화합니다.
        2. 그룹 세션: 대인 관계 기술 및 마음챙김 등의 기술을 그룹 환경에서 연습하며 서로 피드백을 주고받습니다.
        3. 전화 코칭: 환자가 위기 상황에서 치료사에게 전화를 걸어 즉각적인 지도를 받을 수 있도록 합니다.
        4. 치료 팀 회의: 치료사들이 정기적으로 모여 서로의 치료 방법을 논의하고 지원합니다.

    b. CBT (Cognitive Behavioral Therapy: 인지 행동 치료)
    - CBT는 심리 치료에서 매우 널리 사용되는 방법으로, Aaron T. Beck에 의해 개발되었습니다. 이 치료법은 환자의 부정적 사고 패턴을 인식하고 이를 긍정적인 방향으로 변화시켜 감정과 행동을 개선하는 데 중점을 둡니다.

    - CBT의 핵심 구성 요소:
        1. 인지 재구성(Cognitive Restructuring): 왜곡된 사고를 확인하고 이를 도전하여 보다 현실적이고 긍정적인 사고로 바꾸는 기술.
        2. 노출 치료(Exposure Therapy): 두려움이나 회피를 유발하는 상황에 점진적으로 노출하여 불안감을 줄이는 방법.
        3. 행동 활성화(Behavioral Activation): 긍정적인 행동을 촉진하기 위한 전략을 사용하여 우울증을 극복하는 방법.
        4. 자기 관찰(Self-Monitoring): 환자가 자신의 생각, 감정, 행동을 기록하고 이를 통해 패턴을 인식하는 과정.

    - CBT 상담 지침:
        1. 목표 설정: 초기 세션에서 구체적이고 측정 가능한 목표를 설정합니다.
        2. 인지 왜곡 확인: 환자의 부정적이거나 비합리적인 사고를 탐색하고 이를 논리적으로 반박합니다.
        3. 숙제 제공: 세션 외 시간에 환자가 연습할 수 있도록 과제를 주고, 다음 세션에서 피드백을 줍니다.
        4. 점진적 노출: 환자가 두려워하는 상황에 점진적으로 노출되도록 하여 회피 행동을 줄입니다.
        5. 기록 유지: 환자가 자신의 진전을 기록하고 자기 인식을 높일 수 있도록 지도합니다.

    c. 정리
    - DBT는 주로 감정 조절이 어려운 환자들에게 효과적이며, 마음챙김과 같은 기술을 통해 환자가 자신의 감정을 이해하고 관리할 수 있도록 돕습니다.
    - CBT는 부정적인 사고 패턴을 개선하는 데 중점을 두며, 우울증, 불안 장애 등 다양한 심리적 문제에 널리 사용됩니다.
    - 두 치료법 모두 구조화된 접근 방식을 사용하여 환자들이 실질적인 변화를 이룰 수 있도록 돕는 것을 목표로 합니다. 각 방법을 사용하여 상담할 때는 환자의 특성과 필요에 따라 유연하게 접근해야 하며, 지속적인 피드백과 조정이 중요합니다.
"""
completion = client.chat.completions.create(
  model="gpt-4o-mini-2024-07-18",
  messages=[
    {"role": "system", "content": system_msg},
    {"role": "user", "content": prompt}
  ],
  temperature=0.7,          ## 창의적이거나 무작위적인지를 조절: 0에 가까울수록 모델은 더 결정적이고 예측 가능한 출력, 1에 가까울수록 더 무작위적이고 다양한 응답을 생성
  max_tokens=500,            ## 모델이 생성할 수 있는 최대 토큰 수를 지정
  top_p=0.7,                  ## top_p=1이면, 모델은 확률이 가장 높은 단어들을 모두 고려하여 응답을 생성, top_p 값이 낮아지면, 모델은 더 적은 수의 단어들만 고려하게 되어 응답이 더 집중되지만 덜 창의적
  frequency_penalty=0.5,      ## 모델이 동일한 단어를 반복 사용에 벌점: 0이면 벌점이 없고, 모델이 동일한 단어를 반복 사용, 값이 높아지면 동일한 단어를 반복 가능성 감소
  presence_penalty=0.5        ## 새로운 단어를 도입할지를 결정. 값이 높아지면, 모델은 새로운 단어를 도입하는 것을 더 장려
)

print(completion.choices[0].message.content)
```
