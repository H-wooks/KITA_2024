Correlation
=$\frac{Cov}{STD:NFET*STD:pFET}$

==> Cov=$ Correlation \times STD_nFET \times STD_pFET$

#### Covariance Matrix

NFET Ion: Mean = 600, Standard Deviation = 100
PFET Ion: Mean = 500, Standard Deviation = 150 Îùº ÎëêÎ©¥,<br>


Covariance Matrix =
\begin{equation}
   \begin{bmatrix}
   ùúé^2(nFET) & Cov(nFET:pFET)  \\
   Cov(nFET:pFET) & ùúé^2(pFET)  \\
   \end{bmatrix}
\end{equation}

, where Cov(nFET:pFET)=$ Correlation \times ùúé(nFET) \times ùúé(pFET)$ <br>
<br><br><br>

ùúé(nFET)=$100^2$=10000 <br>
ùúé(pFET)=$150^2$=22500 <br>

Covariance Matrix =
\begin{equation}
   \begin{bmatrix}
   10000 & Correlation\times100\times150   \\
   Correlation\times100\times150 & 22500  \\
   \end{bmatrix}
\end{equation}
<br><br><br>
#### Eigen Value & Eigen Vector
Covariance Matrix =
\begin{equation}
   \begin{bmatrix}
   C_{11} & C_{12}  \\
   C_{21} & C_{22}  \\
   \end{bmatrix}
\end{equation} Îùº ÎëêÎ©¥, <br>
<br>
<br>

##### [Eigen Value]

Eigen-values=$Œª_{1},Œª_{2} $

det(Œ£‚àíŒªI)=0 <br>
==> $(Œª‚àíC_{11}) \times (Œª‚àíC_{22})-C_{12} \times C_{21} = 0$ ÏãùÏùÑ ÎßåÏ°±ÌïòÎäî $Œª_{1},Œª_{2} $ <br>

```python
#### Final Version
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Ellipse

np.random.seed(0)

## Group-1
# CMOS Characteristics On-Currents
# Only Random Components
NFET_Ion_1 = np.random.randn(100000) * 100 + 600
PFET_Ion_1 = np.random.randn(100000) * 150 + 500


## Group-2
# Implementing Systematic Variation Components Using Co-Variance
# ÌèâÍ∑† Î∞è ÌëúÏ§ÄÌé∏Ï∞®
mean = [600, 500]     # Mean values of nFET and pFET
std_dev = [100, 150]  # Standard Deviations of nFET and pFET
correlation = 0.5  # ÏõêÌïòÎäî ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ
covariance = correlation * std_dev[0] * std_dev[1]
cov_matrix = [[std_dev[0]**2, covariance], [covariance, std_dev[1]**2]]
# Îã§Î≥ÄÎüâ Ï†ïÍ∑ú Î∂ÑÌè¨ÏóêÏÑú ÏÉòÌîå ÏÉùÏÑ±
FET_Ion = np.random.multivariate_normal(mean, cov_matrix, size=100000)

## Group-3
random_nFET = np.random.randn(100000)  # Un-Correlated
random_pFET = np.random.randn(100000)  # Un-Correlated
systematic_variation = np.random.randn(100000)  # Systematic Variation

NFET_Ion_3 = (random_nFET + systematic_variation)*100 + 600
PFET_Ion_3 = (random_pFET + systematic_variation)*150 + 500


# Group-4 with specified correlation coefficient
# Generate uncorrelated random variables

mean4 = [600, 500]     # Mean values of nFET and pFET
std_dev4 = [100, 150]  # Standard Deviations of nFET and pFET
random_nFET4 = np.random.randn(100000)
random_pFET4 = np.random.randn(100000)

# Introduce correlation
correlation4 = 0.5
random_pFET4 = correlation4 * random_nFET4 + np.sqrt(1 - correlation4**2) * random_pFET4
# Systematic variation
systematic_variation4 = np.random.randn(100000)

# Create final variables with specified means and std deviations
NFET_Ion_4 = (random_nFET4 * std_dev4[0] + mean4[0]) + systematic_variation4
PFET_Ion_4 = (random_pFET4 * std_dev4[1] + mean4[1]) + systematic_variation4

DF_CMOS = pd.DataFrame({
    'nFET-1': NFET_Ion_1, 'pFET-1': PFET_Ion_1,
    'nFET-2': FET_Ion[:,0], 'pFET-2': FET_Ion[:,1],
    'nFET-3': NFET_Ion_3, 'pFET-3': PFET_Ion_3,
    'nFET-4': NFET_Ion_4, 'pFET-4': PFET_Ion_4
})

# Scatter plot ÏÉùÏÑ±
fig, axes = plt.subplots(2, 2, figsize=(10, 7))

# Ï≤´ Î≤àÏß∏ subplot
ax1 = axes[0,0]
sns.scatterplot(x=DF_CMOS['nFET-1'], y=DF_CMOS['pFET-1'], alpha=0.1, color='c', label='Group-1', ax=ax1, s=5)
ax1.set_title('CMOS Ion Distribution n-/p-FET Correlation')
ax1.set_xlabel("nFET's Ion")
ax1.set_ylabel("pFET's Ion")
ax1.legend()

# Îëê Î≤àÏß∏ subplot
ax2 = axes[0,1]
sns.scatterplot(x=DF_CMOS['nFET-2'], y=DF_CMOS['pFET-2'], alpha=0.1, color='b', label='Group-2', ax=ax2, s=5)
ax2.set_title('CMOS Ion Distribution n-/p-FET Correlation')
ax2.set_xlabel("nFET's Ion")
ax2.set_ylabel("pFET's Ion")
ax2.legend()

# ÏÑ∏ Î≤àÏß∏ subplot
ax3 = axes[1,0]
sns.scatterplot(x=DF_CMOS['nFET-3'], y=DF_CMOS['pFET-3'], alpha=0.1, color='m', label='Group-3', ax=ax3, s=5)
ax3.set_title('CMOS Ion Distribution n-/p-FET Correlation')
ax3.set_xlabel("nFET's Ion")
ax3.set_ylabel("pFET's Ion")
ax3.legend()

# ÎÑ§ Î≤àÏß∏ subplot
ax4 = axes[1,1]
sns.scatterplot(x=DF_CMOS['nFET-4'], y=DF_CMOS['pFET-4'], alpha=0.1, color='g', label='Group-4', ax=ax4, s=5)
ax4.set_title('CMOS Ion Distribution n-/p-FET Correlation')
ax4.set_xlabel("nFET's Ion")
ax4.set_ylabel("pFET's Ion")
ax4.legend()

# Median Î∞è Mean Í≥ÑÏÇ∞
median_1 = DF_CMOS['nFET-1'].median(), DF_CMOS['pFET-1'].median()
median_2 = DF_CMOS['nFET-2'].median(), DF_CMOS['pFET-2'].median()
median_3 = DF_CMOS['nFET-3'].median(), DF_CMOS['pFET-3'].median()
median_4 = DF_CMOS['nFET-4'].median(), DF_CMOS['pFET-4'].median()

mean_1 = DF_CMOS['nFET-1'].mean(), DF_CMOS['pFET-1'].mean()
mean_2 = DF_CMOS['nFET-2'].mean(), DF_CMOS['pFET-2'].mean()
mean_3 = DF_CMOS['nFET-3'].mean(), DF_CMOS['pFET-3'].mean()
mean_4 = DF_CMOS['nFET-4'].mean(), DF_CMOS['pFET-4'].mean()

sigma3_1 = 3*DF_CMOS['nFET-1'].std(), 3*DF_CMOS['pFET-1'].std()
sigma3_2 = 3*DF_CMOS['nFET-2'].std(), 3*DF_CMOS['pFET-2'].std()
sigma3_3 = 3*DF_CMOS['nFET-3'].std(), 3*DF_CMOS['pFET-3'].std()
sigma3_4 = 3*DF_CMOS['nFET-4'].std(), 3*DF_CMOS['pFET-4'].std()

# Deviation Contour Plot
def plot_3sigma(ax, mean, cov, sigma3, color):
    eigenvalues, eigenvectors = np.linalg.eigh(cov)
    order = eigenvalues.argsort()[::-1]
    eigenvalues = eigenvalues[order]
    eigenvectors = eigenvectors[:, order]  # eigenvectorsÏùò ÏàúÏÑúÎ•º Î≥ÄÍ≤Ω
    scaling_factor=2.3
    std_deviation = np.sqrt(eigenvalues) * 3

    ### Corner Points
    FF = (mean[0] + sigma3[0], mean[1] + sigma3[1])
    SS = (mean[0] - sigma3[0], mean[1] - sigma3[1])
    FSE = (mean[0] + sigma3[0], mean[1] - sigma3[1])
    FS = (mean[0] + scaling_factor/3*sigma3[0], mean[1] - scaling_factor/3*sigma3[1])
    # FS = mean - 2.5 * std_deviation[0] * eigenvectors[:, 1]

    SFE = (mean[0] - sigma3[0], mean[1] + sigma3[1])
    SF = (mean[0] - scaling_factor/3*sigma3[0], mean[1] + scaling_factor/3*sigma3[1])
    # SF = mean + 2.5 * std_deviation[0] * eigenvectors[:, 1]

    NN = (mean[0], mean[1])

    ellipse = Ellipse(xy=mean, width=std_deviation[0] * 2, height=std_deviation[1] * 2,
                      angle=np.degrees(np.arctan2(*eigenvectors[:, 0][::-1])),
                      edgecolor=color, fc='None', lw=1, linestyle='--')
    ax.add_patch(ellipse)

    # NN, FF, SS, FS, SF Corner Points Plot
    ax.plot(FF[0], FF[1], 'o', color='red')
    ax.text(FF[0], FF[1], 'FF', fontsize=12, ha='right', va='bottom')

    ax.plot(SS[0], SS[1], 'o', color='blue')
    ax.text(SS[0], SS[1], 'SS', fontsize=12, ha='right', va='top')

    ax.plot(FS[0], FS[1], 'o', color='green')
    ax.text(FS[0], FS[1], 'FS', fontsize=12, ha='right', va='top')

    ax.plot(FSE[0], FSE[1], 'o', color='green')
    ax.text(FSE[0], FSE[1], 'FSE', fontsize=12, ha='right', va='top')

    ax.plot(SF[0], SF[1], 'o', color='purple')
    ax.text(SF[0], SF[1], 'SF', fontsize=12, ha='left', va='bottom')

    ax.plot(SFE[0], SFE[1], 'o', color='purple')
    ax.text(SFE[0], SFE[1], 'SFE', fontsize=12, ha='left', va='bottom')

    # Nominal Corner Plot
    ax.plot(NN[0], NN[1], 'o', color='black')
    ax.text(NN[0], NN[1], 'NN', fontsize=12, ha='right', va='bottom')

    # Corner Box Plot
    rect_x = [FF[0], SFE[0], SS[0], FSE[0], FF[0]]
    rect_y = [FF[1], SFE[1], SS[1], FSE[1], FF[1]]
    ax.plot(rect_x, rect_y, linestyle='--', color='gray', linewidth=0.4)

    # Plot eigen-vector lines
    ax.plot([mean[0], SFE[0]], [mean[1], SFE[1]], color='k', linestyle='--', linewidth=0.2)
    ax.plot([mean[0], FSE[0]], [mean[1], FSE[1]], color='k', linestyle='--', linewidth=0.2)
    ax.plot([mean[0], FF[0]], [mean[1], FF[1]], color='k', linestyle='--', linewidth=0.2)
    ax.plot([mean[0], SS[0]], [mean[1], SS[1]], color='k', linestyle='--', linewidth=0.2)


# 3ÏãúÍ∑∏Îßà Î≤îÏúÑ Í≥ÑÏÇ∞ Î∞è Í∑∏Î¶¨Í∏∞
cov_1 = np.cov(DF_CMOS[['nFET-1', 'pFET-1']], rowvar=False)
cov_2 = np.cov(DF_CMOS[['nFET-2', 'pFET-2']], rowvar=False)
cov_3 = np.cov(DF_CMOS[['nFET-3', 'pFET-3']], rowvar=False)
cov_4 = np.cov(DF_CMOS[['nFET-4', 'pFET-4']], rowvar=False)

plot_3sigma(ax1, mean_1, cov_1, sigma3_1, 'c')  # Ï≤´ Î≤àÏß∏ subplotÏóê ÌÉÄÏõê Í∑∏Î¶¨Í∏∞
plot_3sigma(ax2, mean_2, cov_2, sigma3_2, 'b')  # Îëê Î≤àÏß∏ subplotÏóê ÌÉÄÏõê Í∑∏Î¶¨Í∏∞
plot_3sigma(ax3, mean_3, cov_3, sigma3_3, 'm')  # ÏÑ∏ Î≤àÏß∏ subplotÏóê ÌÉÄÏõê Í∑∏Î¶¨Í∏∞
plot_3sigma(ax4, mean_4, cov_4, sigma3_4, 'b')  # ÎÑ§ Î≤àÏß∏ subplotÏóê ÌÉÄÏõê Í∑∏Î¶¨Í∏∞

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Ellipse

np.random.seed(0)
################# CMOS Characteristics On-Currents
NFET_Ion_1=np.random.normal(loc=600, scale=100, size=100000)
PFET_Ion_1=np.random.normal(loc=500, scale=150, size=100000)

################# CMOS Characteristics On-Currents (randn)
NFET_Ion_2=np.random.randn(100000)*100+600
PFET_Ion_2=np.random.randn(100000)*150+500

DF_CMOS = pd.DataFrame({
    'nFET-1': NFET_Ion_1, 'pFET-1': PFET_Ion_1,
    'nFET-2' : NFET_Ion_2, 'pFET-2': PFET_Ion_2
})

# Scatter plot ÏÉùÏÑ±
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
ax1 = plt.gca()  # Ï≤´ Î≤àÏß∏ subplotÏùò Ï∂ï Í∞ÄÏ†∏Ïò§Í∏∞
sns.scatterplot(x=DF_CMOS['nFET-1'] , y=DF_CMOS['pFET-1'], alpha=0.1, color='c', label='Group-1')
plt.title('CMOS Ion Distribution n-/p-FET Correlation')
plt.xlabel("nFET's Ion")
plt.ylabel("pFET's Ion")
plt.legend()

plt.subplot(1, 2, 2)
ax2 = plt.gca()  # Ï≤´ Î≤àÏß∏ subplotÏùò Ï∂ï Í∞ÄÏ†∏Ïò§Í∏∞
sns.scatterplot(x=DF_CMOS['nFET-2'] , y=DF_CMOS['pFET-2'], alpha=0.1, color='b', label='Group-2')
plt.title('CMOS Ion Distribution n-/p-FET Correlation')
plt.xlabel("nFET's Ion")
plt.ylabel("pFET's Ion")
plt.legend()

# Median Î∞è Mean Í≥ÑÏÇ∞
median_1 = DF_CMOS['nFET-1'].median(), DF_CMOS['pFET-1'].median()
median_2 = DF_CMOS['nFET-2'].median(), DF_CMOS['pFET-2'].median()

mean_1 = DF_CMOS['nFET-1'].mean(), DF_CMOS['pFET-1'].mean()
mean_2 = DF_CMOS['nFET-2'].mean(), DF_CMOS['pFET-2'].mean()

# 3ÏãúÍ∑∏Îßà Î≤îÏúÑÏóê Ìï¥ÎãπÌïòÎäî ÌÉÄÏõê Í∑∏Î¶¨Í∏∞
def plot_3sigma(ax, mean, cov, color):
    sigma_x = np.sqrt(cov[0, 0])
    sigma_y = np.sqrt(cov[1, 1])
    sigma_xy = cov[0, 1]

    theta = np.arctan2(2 * sigma_xy, (sigma_x**2 - sigma_y**2)) / 2

    width = 2 * np.sqrt(2 * np.log(2) * (sigma_x**2 + sigma_y**2 +
                                         np.sqrt((sigma_x**2 - sigma_y**2)**2 + (2*sigma_xy)**2)))
    height = 2 * np.sqrt(2 * np.log(2) * (sigma_x**2 + sigma_y**2 -
                                          np.sqrt((sigma_x**2 - sigma_y**2)**2 + (2*sigma_xy)**2)))
    ellipse = Ellipse(xy=mean, width=width, height=height, angle=np.degrees(theta),
                      edgecolor=color, fc='None', lw=2)
    ax.add_patch(ellipse)

# 3ÏãúÍ∑∏Îßà Î≤îÏúÑ Í≥ÑÏÇ∞ Î∞è Í∑∏Î¶¨Í∏∞
cov_1 = np.cov(DF_CMOS[['nFET-1', 'pFET-1']], rowvar=False)
cov_2 = np.cov(DF_CMOS[['nFET-2', 'pFET-2']], rowvar=False)

plot_3sigma(ax1, mean_1, cov_1, 'red')  # Ï≤´ Î≤àÏß∏ subplotÏóê ÌÉÄÏõê Í∑∏Î¶¨Í∏∞
plot_3sigma(ax2, mean_2, cov_2, 'green') # Îëê Î≤àÏß∏ subplotÏóê ÌÉÄÏõê Í∑∏Î¶¨Í∏∞

plt.tight_layout()
plt.show()

```
