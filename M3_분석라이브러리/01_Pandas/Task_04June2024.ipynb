Task6. 0일부터 10일까지의 시간 범위를 12시간 간격으로 생성하고, 각 간격에 대해 누적 시간을 계산하여 데이터프레임을 작성하세요.

```python
# 시간 간격 범위 생성
time_delta_range = pd.timedelta_range(start='0', end='100 days', freq='12H')
time_delta_range
```

```python
# 데이터프레임 생성
df = pd.DataFrame({
    'timedelta': time_delta_range,
    'cumulated_time': [timedelta.total_seconds()/3600 for timedelta in time_delta_range]
})

# 데이터프레임 출력
print(df.head(30))
```

Task8. 주어진 df에서 월별로 데이터의 합계를 계산하세요.


```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-15', '2022-02-01', '2022-02-15', '2022-03-01'],
    'value': [1, 2, 3, 4, 5]
})
df
```

```python
# 'date' 열을 datetime 형식으로 변환
df['date'] = pd.to_datetime(df['date'])

# 'month' 열을 추가하여 월 단위로 그룹화
df['month'] = df['date'].dt.to_period('M')

# 월별 합산
monthly_sum = df.groupby('month')['value'].sum().reset_index()

print(monthly_sum)
```

Task9. 주어진 데이터프레임에서 주말(토요일과 일요일) 데이터를 필터링하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-08', '2022-01-09'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df['day of week']=df['date'].dt.dayofweek
df.head()
```

```python
weekend_df=df[df['day of week'] >= 5]
weekend_df
```

Task10. 주어진 df에서 분기별 데이터의 평균을 계산하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])

df
```
