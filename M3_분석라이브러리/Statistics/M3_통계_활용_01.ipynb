Q. 한 학급의 학생 10명의 수학 시험 점수는 다음과 같습니다:

[85, 72, 90, 68, 88, 76, 95, 89, 77, 84].

학급의 평균 점수와 표준편차를 계산하여 전체 성적 수준과 변동성을 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Math=np.array([85, 72, 90, 68, 88, 76, 95, 89, 77, 84])
print("학급의 평균점수:", np.mean(Math), "\n")
print("학급의 중간점수:", np.median(Math), "\n")
print("학급의 시험점수의 표준 편차: ",round(np.std(Math),2), "\n")
print(np.percentile(Math, [25, 50, 75]))

df = pd.DataFrame([85, 72, 90, 68, 88, 76, 95, 89, 77, 84], columns=['Math'])
print(df.describe())
df['Score_Distr']= pd.cut(df['Math'], bins=[x for x in range(0, 101, 10)], labels=[x for x in range(10, 101, 10)])
sns.countplot(x='Score_Distr', data=df)
plt.title('Score Distribution')
plt.show()
###################################################### Insight
## 학급의 평균은 82점이고, 편차가 8점 정도로 70점 이상 상위권에 고루 분포하고, 80~90점대가 가장 많이 포진
```

Q. 한 온라인 쇼핑몰의 지난 10일간 일일 방문자 수는 다음과 같습니다:

[150, 200, 220, 180, 190, 170, 210, 230, 160, 175].

Q. 한 회사의 직원 연봉 데이터가 다음과 같습니다:

[45000, 50000, 52000, 47000, 48000, 53000, 60000, 62000, 55000, 58000].

이 데이터를 이용하여 연봉의 분포와 이상치를 시각화하는 상자 그림을 그리고, 연봉 분포에 대한 인사이트를 도출하세요.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Salary=pd.DataFrame(data=[45000, 50000, 52000, 47000, 48000, 53000, 60000, 62000, 55000, 58000], columns=['Indv Salary'])
print(np.percentile(Salary['Indv Salary'], [25, 50, 75]))

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

Salary['Indv Salary'].plot(kind='bar', ax=axes[0])
axes[0].set_xlabel('Person')
axes[0].set_ylabel('Indv Salary')

Salary.plot.box(ax=axes[1])
axes[1].set_title("Individual Salary")

plt.tight_layout()
plt.show()





```

```python
## 강사님 답안
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Salary=pd.DataFrame(data=[45000, 50000, 52000, 47000, 48000, 53000, 60000, 62000, 55000, 58000], columns=['Indv Salary'])
perc=np.percentile(Salary['Indv Salary'], [25, 50, 75])
print("percentile:", perc)
q1_salary=perc[0]
q3_salary=perc[2]
iqr_salary=perc[2]-perc[0]
print(f"Q1: {q1_salary}, Q3: {q3_salary}, IQR: {iqr_salary}")

### 이상치 확인
lower_bound=q1_salary-1.5*iqr_salary
upper_bound=q3_salary+1.5*iqr_salary
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
outliers=Salary[(Salary['Indv Salary']<lower_bound)|(Salary['Indv Salary']>upper_bound)]
print("Outliers: ",outliers)



fig, axes = plt.subplots(1, 2, figsize=(12, 5))

Salary['Indv Salary'].plot(kind='bar', ax=axes[0])
axes[0].set_xlabel('Person')
axes[0].set_ylabel('Indv Salary')

Salary.plot.box(ax=axes[1])
axes[1].set_title("Individual Salary")

plt.tight_layout()
plt.show()

```

Q. 한 마케팅 캠페인에서 투자한 광고비와 발생한 매출 데이터가 각각 다음과 같습니다:

광고비 [1, 2, 3, 4, 5] (단위: 천 달러),

매출 [10, 15, 20, 25, 30] (단위: 천 달러).

광고비와 매출 간의 관계를 시각화하기 위해 산점도를 그리고, 두 변수 간의 관계에 대한 인사이트를 도출하세요.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dict_data = {'광고비':[1,2,3,4,5], '매출':[10, 15, 20, 25, 30]}
df = pd.DataFrame(dict_data)

coeff = np.polyfit(df['광고비'], df['매출'], 1)  # 1=linear
m = coeff[0]
b = coeff[1]

correlation=np.corrcoef(df['광고비'], df['매출'])
print(correlation)
correlation1=np.corrcoef(df['광고비'], df['매출'])[0,1]
print(correlation1)

print(f"기울기:{m}, 절편: {b}")

plt.figure(figsize=(6, 4))
df.plot.scatter(x='광고비', y='매출')
plt.plot(df['광고비'], m*df['광고비']+b)
plt.xlabel('Advertising expenditure')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()



```

Q. 한 식당의 월별 홍보 비용과 월별 방문자 수 데이터가 각각 다음과 같습니다:

홍보 비용 [500, 600, 700, 800, 900] (단위: 달러),

방문자 수 [300, 400, 500, 600, 700] (단위: 명).

홍보 비용과 방문자 수 간의 상관 계수를 계산하여 관계를 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dict_data = {'홍보비':[500, 600, 700, 800, 900], '방문자수':[300, 400, 500, 600, 700]}
df = pd.DataFrame(dict_data)
coeff = np.polyfit(df['홍보비'], df['방문자수'], 1)  # 1=linear
m = coeff[0]
b = coeff[1]
print(f"기울기:{m}, 절편: {b}")
correlation1=np.corrcoef(df['홍보비'], df['방문자수'])[0,1]
print(correlation1)
df.plot.scatter(x='홍보비', y='방문자수')
plt.plot(df['홍보비'], m*df['홍보비']+b)
plt.xlabel('Advertising expenditure')
plt.ylabel('No_Visitors')
plt.show()
```

```python
ad_cost=np.array([500, 600, 700, 800, 900])
visitors=np.array([300, 400, 500, 600, 700])
v_count=np.sort(visitors)[::-1]
correlation=np.corrcoef(ad_cost, v_count)[0,1]
print(correlation)

###
if correlation > 0.7:
  print("홍보 비용과 방문자 수간의 상관 관계가 높습니다.")
elif correlation > 0.5:
  print("홍보 비용과 방문자 수간의 상관 관계가 중간입니다.")
else:
  print("홍보 비용과 방문자 수간의 상관 관계가 낮거나 없습니다.")
```

Q. 세 종류의 화합물에 대한 실험 데이터를 다음 매트릭스로 나타냅니다:

Q. 한 회사의 연구개발(R&D) 투자 비용과 연간 수익 데이터가 각각 다음과 같습니다:

투자 비용 [100, 200, 300, 400, 500] (단위: 천 달러),

연간 수익 [200, 400, 600, 800, 1000] (단위: 천 달러).

추세선이 있는 산점도를 그려 투자 비용과 연간 수익 간의 관계를 시각화하고, 인사이트를 도출하세요.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
dict_data = {'Investment':[100, 200, 300, 400, 500], 'Profits':[200, 400, 600, 800, 1000]}
df = pd.DataFrame(dict_data)
df.plot.scatter(x='Investment', y='Profits')
m, b = np.polyfit(df['Investment'], df['Profits'], 1)  # 1=linear

print(f"기울기:{m}, 절편: {b}")
plt.plot(df['Investment'], m*df['Investment']+b)
plt.title('Profits vs. Investment')
plt.xlabel('Investment')
plt.ylabel('Profits')
plt.tight_layout()
plt.show()
```
