Task1_0617. 선그래프, 히스토그램, 산점도, 막대그래프 4개의 subplot으로 구성된 2 x 2 그래프를 작성하세요.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

filepath='/content/drive/MyDrive/KITA_2024/M3_분석라이브러리/Pandas/dataset/auto-mpg.xlsx'
df=pd.read_excel(filepath, engine='openpyxl')

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].plot(df['weight'], df['mpg'])
axes[0, 0].set_title('Line Plot (MPG vs. Weight)')
axes[0, 0].set_xlabel('Weight')
axes[0, 0].set_ylabel('MPG')

# 두 번째 서브플롯
df['mpg'].plot(kind='hist', bins=30, ax=axes[0,1], title='Histogram of MPG')
#axes[0, 1].set_title('Line Plot (MPG vs. Acceleration)')
axes[0, 1].set_xlabel('MPG')
axes[0, 1].set_ylabel('Freq')

# 세 번째 서브플롯
axes[1, 0].plot(df['cylinders'], df['mpg'])
axes[1, 0].set_title('Line Plot (MPG vs. Horsepower)')
axes[1, 0].set_xlabel('Horsepower')
axes[1, 0].set_ylabel('MPG')

# 네 번째 서브플롯
axes[1, 1].plot(df['cylinders'], df['mpg'])
axes[1, 1].set_title('Line Plot (MPG vs. Cylinders)')
axes[1, 1].set_xlabel('Cylinders')
axes[1, 1].set_ylabel('MPG')

plt.tight_layout()
plt.show()

# df['A'].plot(, title='Line Plot')                                                   ## Line
# df['A'].plot(kind='hist', bins=30, ax=axes[0,1], title='Histogram')                             ## 히스토그램
# df['A'].plot(kind='box', ax=axes[0,2], title='box Plot')                                        ## 박스
# df.plot(kind='scatter', x='A', y='B', ax=axes[1,0], title='Scatter Plot')                       ## 산점도
# df[['A','B']].cumsum().plot(kind='area', ax=axes[1,1], title='Area Plot', stacked=False)        ## 면적그래프 (stacked =False는 두 시리즈가 겹치지 않도록 하는 옵션)
# df['D'].plot(kind='bar', ax=axes[1,2], title='Bar Plot')                                        ## 막대 그래프
# df['D'].value_counts().head(10).plot(kind='pie', ax=axes[2,0], title='Pie Chart')               ## 파이 차트 (상위 10개만 plot)
# df['A'].plot(kind='kde', ax=axes[2,1], title='KDE Plot')                                        ## 커널 밀도 추정 그래프
# df['D'].value_counts().head(10).plot(kind='barh', ax=axes[2,2], title='Horizontal Bar Plot')    ## 수평 막대 그래프








```
