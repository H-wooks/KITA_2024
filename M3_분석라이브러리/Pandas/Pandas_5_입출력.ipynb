```python
# 저장할 때
import numpy as np
import pandas as pd
ar1 = np.arange(10).reshape(2,5)
df = pd.DataFrame(ar1,columns=list('abcde'))
df.to_csv('test.csv', index=None)
```

```python
df1=df.copy()
df1.to_csv('test1.csv')
df1=pd.read_csv('test1.csv', index_col=0)
df1
```

```python
df1.iloc[0]=2
df1.iloc[1]=3
df1
```

```python
# Json 파일 읽고 쓰기
df1.to_json('test2.json')
df_j=pd.read_json('test2.json')
df_j
```

```python
# 바이너리 파일 읽고 쓰기
df1.to_pickle('test3.pkl')
df1=pd.read_pickle('test3.pkl')
df1
```
