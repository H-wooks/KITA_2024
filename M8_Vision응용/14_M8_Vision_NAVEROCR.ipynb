#### Python에서 환경 변수 PATH에 새로운 경로를 추가하는 방식

1. os.environ["PATH"]:
- Python의 os 모듈을 사용하여 환경 변수를 다룰 수 있는데, os.environ은 현재 시스템의 환경 변수를 담고 있는 딕셔너리
- 그 중 PATH는 운영 체제가 실행 파일을 찾는 경로 목록

<br>

2. += os.pathsep + "/usr/bin":
- += 연산자는 PATH 환경 변수에 새로운 값을 추가하는 역할을 합니다.
- os.pathsep은 운영 체제에 따라 경로를 구분하는 문자입니다. 예를 들어, 유닉스 계열에서는 :이고, 윈도우에서는 ;입니다.
- 여기에 /usr/bin 경로를 추가하여, /usr/bin에 있는 실행 파일도 시스템에서 찾을 수 있도록 합니다.

###### 결과적으로, 이 코드는 시스템의 PATH 환경 변수에 /usr/bin 경로를 추가하여, Python이 해당 경로에 있는 프로그램을 실행할 수 있도록 만드는 것입니다.

```python
import os
os.environ["PATH"] += os.pathsep + "/usr/bin"
```

```python
from pdf2image import convert_from_path
import os

pdf_path='/content/text_eng.pdf'
images=convert_from_path(pdf_path, dpi=300)

for i, image in enumerate(images):
  plt.figure(figsize=(16,10))
  plt.imshow(image)
  plt.axis('off')
  plt.title(f"Page{i+1}")
  plt.show()
  #image.save(f'page{i+1}.jpg', 'JPEG')
```

```python
import cv2
import pytesseract
from pdf2image import convert_from_path
from PIL import Image
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import os
import numpy as np

# PDF 파일을 이미지로 변환
pages = convert_from_path('/content/text_eng.pdf')

# 텍스트를 저장할 변수 초기화
text = ""

# 이미지를 그레이스케일로 변환하고 텍스트 추출
for idx, page in enumerate(pages):
    # 이미지를 그레이스케일로 변환
    gray = cv2.cvtColor(np.array(page), cv2.COLOR_RGB2GRAY)

    # 이미지를 파일로 저장하여 pytesseract로 텍스트 추출
    filename = "{}.jpg".format(idx)
    cv2.imwrite(filename, gray)
    text += pytesseract.image_to_string(Image.open(filename), lang='eng')
    os.remove(filename)

# 추출된 텍스트 출력
print(text)
```

```python
from pdf2image import convert_from_path
import pytesseract
from PIL import Image
from PIL import ImageEnhance, ImageFilter
import re

images = convert_from_path('/content/text_ko.pdf', dpi=300)

# 변환된 이미지로부터 텍스트 추출 및 전처리
all_text = ''
for image in images:
    # 이미지 전처리: 그레이스케일로 변환, 대비 조정, 노이즈 제거
    image = image.convert('L')  # 그레이스케일로 변환
    enhancer = ImageEnhance.Contrast(image)
    image = enhancer.enhance(2)  # 대비를 2배로 조정
    image = image.filter(ImageFilter.MedianFilter(size=3))  # 노이즈 제거

    # pytesseract를 사용하여 이미지에서 텍스트 추출
    text = pytesseract.image_to_string(image, lang='eng+kor')

    all_text += text

# 추출된 텍스트 출력 (또는 다른 처리를 위해 사용)
print(all_text)
```
