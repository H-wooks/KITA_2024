#### 18. 사칙연산을 수행하는  클래스를 작성하여 객체를 만들고 2개의 정수 10, 5를 입력한 후 연산 결과를  포멧을 이용하여 출력하세요.

```python
class FundArith:
  a=int(input('사칙연산을 위한 첫번째 수 입력:> '))
  b=int(input('사칙연산을 위한 두번째 수 입력:> '))
  def Cal(self,operator):
    if self.a == 0:
      self.a=int(input('올바른 수를 입력하세요 (첫번째 수):> '))
    if self.b == 0:
      self.b=int(input('올바른 수를 입력하세요 (첫번째 수):> '))
    arith_eq="self.a"+operator+"self.b"
    print("{} {} {} = {}".format(self.a,operator,self.b,eval(arith_eq)))


obj=FundArith()
obj.Cal("+")
obj.Cal("-")
obj.Cal("*")
obj.Cal("/")
```

#### 20. 문자열 리스트 states를 아래 [출력]와 같이 전처리하여 출력하세요
`states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina##', 'West virginia?']`

[출력]\
['Alabama',\
 'Georgia',\
 'Georgia',\
 'Georgia',\
 'Florida',\
 'South Carolina',\
 'West Virginia']

```python
import re

states = [' Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda', 'south carolina##', 'West virginia?']
result=[]
for word in states:
  word=re.sub(r'[ ?.,!#\n\';:]', '', word).title()
  result.append(word)

print(result)
```
