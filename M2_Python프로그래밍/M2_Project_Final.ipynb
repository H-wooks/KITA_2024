```python



deletingEntry=['the', 'when', 'two', 'twenty-eight', 'something', 'beside', 'want', 'eldest', 'of', 'near', 'next', 'mother', 'both', 'hands', 'eyes', 'have-this', 'like', 'for', 'son', 'some','bedroom','mind','heavy','misshapen','where','side','my','form','appeared','are','seemed','castel','horse','mountain','brother','world','little','four','aedificium','first','door','here','then','emmett','away','nightfifteen','urgentone','still-shocked','looking','minutes',
               '1911', 'legs', 'too', 'this','his', 'help', 'but', 'from', 'couldnt', 'wrong', 'girls','after', 'with','sons','washed', 'walk', 'lost', 'again', 'about', 'knew', 'never', 'a', 'extra','place','s','ll'
               'at', 'just', 'that', 'her', 'having', 'old', 'middle', 'read', 'same', 'cold', 'one', 'did', 'foot;',  'korea', 'long', 'moved', 'kitchen;', 'didnt', 'happens', 'weak—the', 'orphans—were',
               'well', 'late', 'who', 'run', 'sleeping',  'an', 'korean', 'hoonies', 'aging', 'make', 'over', 'yeongdo', 'as','three', 'golden', 'made', 'himthe', 'themselves', 'father', 'wife', 'feet',
               'lived', 'on', 'koreahistory', 'work', 'daughter', 'larger', 'full', 'being', 'would', 'mouth', 'him',  'not', 'were', 'she', 'lodgersthe', 'other', 'have', 'sons—', 'way', 'cup', 'head',
               'increase', 'quicklyboth', 'along', 'couple', 'girl', 'no', 'worked', 'whod', 'than', 'him', 'let', 'he', 'habit', 'outdoor', 'keep', 'city', 'front', 'loved', 'twenty-seven', 'life', 'had',
               'needed', 'talk', 'strong', 'ricein', 'evening', 'japanese', 'how','their', 'to', 'could', 'know', 'eventually', 'paper', 'in', 'hot', 'most', 'fathers', 'write', 'however', 'stone', 'close',
               'countries', 'yet', 'survived', 'turn', 'man', 'avoided', 'its', 'benefit', 'wallat', 'seen', 'hundred', 'known', 'finished', 'called', 'generation', 'mild', 'well-shaped', 'share', 'century',
               'sums', 'wooden', 'pointless', 'resembled', 'twice', 'number', 'inky', 'peasants', 'things', 'guessed', 'quickly', 'has', 'rent', 'gave', 'home', 'there', 'up', 'born', 'heart', 'village',
               'snug', 'low', 'child', 'fathom', 'vegetable', 'decided', 'rivals', 'learned', 'someone', 'dining', 'say', 'busan', 'asked', 'steady', 'nimble', 'twisted', 'be', '1910', 'wanted', 'families',
               'thinking', 'because', 'carry', 'out', 'gone', 'birth', 'so', 'tables', 'what', 'i', 'and', 'raised', 'growing', 'stay', 'day', 'morning', 'rooms', 'spring', 'money', 'build', 'through', 'since',
               'elderly', 'years', 'hoonie', 'fisherman', 'us', 'himself', 'slept', 'take', 'do', 'by', 'they',  'possible', 'nature', 'it', 'five-mile-wide', 'house', 'even', 'young', 'more', 'japan', 'age',
               'any', 'doors', 'yeongdo—a', 'only', 'hardy', 'was', 'five', 'youngest', 'kitchen', 'bed', 'such','or','large','parents','lodgers', 'hed', 'weeks', 'we', 'me', 'which', 'those', 'you', 'enough','kept','family','market','school','speak','shoulders','square','port','truein','cheated','organ','fished','sides','see','road','william','said','can','came','will','appreared','if','earth','your','them','they','said','into','going','tear','charlie','all','voice','every','took','always','inside','edward','bella','looked','ran','love','shut','reached','much','room','this',"He's" ]

########## 외부 라이브러리 패키지가 설치 되어 있지 않으면, Package-Installer-of-python로부터 아래 라이브러리 패키지들을 인스톨
########## grep해서 결과가 있을때 종료 코드 0으로 Exit, grep의 결과가 없으면 1을 반환
import os
if os.system("pip list | grep googletrans"):
  !pip install googletrans==4.0.0rc1
if os.system("pip list | grep natsort"):
  !pip install natsort
if os.system("pip list | grep gTTS"):
  !pip install gTTS
if os.system("pip list | grep playsound"):
  !pip install playsound

import googletrans
from gtts import gTTS
import playsound as ps
import re
import pickle
import random
curpath='/content/drive/MyDrive/00_KITA_2404/M2_Program/'
os.chdir(curpath)

########################################################## 초기 가이드를 위한 display에 필요한 문자열
str1="Welcome to Our Translation/Voiceenabled Program"
str2="\t\t용도 및 목적: 외국어 원문 Reading 학습을 위한 번역 및 Listening 학습을 위한 원어 음성 파일 생성 "
str3='\t\t기능:'
str4='\t\t(i) 사용자 입력 TEXT (1. sample-TEXTs, 2. File-Input, 3. Text-Input Window)로 부터 원하는 언어로 번역'
str5='\t\t(ii) 번역된 언어를 음성 파일(MP3)로 변환하여 Listening 반복 학습 기능 제공'
str6='\t\t(iii) 사용자 입력 외국어 원문으로부터 빈출 단어를 추출하여 사용자에게 제공하여 학습 효율 개선'
str7='\t\t(iv) 추출된 빈출 단어를 이용한 퀴즈 서비스 제공하여 사용자의 daily 학습 체크'
langs=['한국어','영어','일본어','중국어-간체','중국어-번체','러시아어','프랑스어','독일어','이탈리아어','스페인','베트남어','태국','인도시아어','체코어','덴마크','그리스어']
langCodes=['ko','en','ja','zh-CN','zh-TW','ru','fr','de','it','es','vi','th','id','cs','da','el']

class TransSpeech:
  textTitle = ""        # 텍스트 제목
  selectedInput = ""    # 텍스트 선택값
  TEXT_DB = {}          # 문서목록

  def InitDisp(self):                                          # 프로그램 소개 및 기능 Display
    print("#"*200,end='');
    for i in range(2): print("\n##"+" "*195,"##",end='')
    print(f'\n##{str1:^195} ##',end='')
    for i in range(2): print("\n##"+" "*195,"##",end='')
    print(f'\n##{str2:<152} ##',end='')
    for i in range(2): print("\n##"+" "*195,"##",end='')
    print(f'\n##{str3:<181} ##',end='')
    print(f'\n##{str4:<167} ##',end='')
    print(f'\n##{str5:<160} ##',end='')
    print(f'\n##{str6:<145} ##',end='')
    print(f'\n##{str7:<155} ##',end='')
    for i in range(2): print("\n##"+" "*195,"##",end='')
    i=1
    for lang in langs:
          if i%8==1: print("\n##  ",end="")
          print(f"{i:2}: {lang:10}",end="")
          if i%8==0: print(" "*53+"##",end="")
          i += 1
    for i in range(4): print("\n##"+" "*195,"##",end='')
    print('\n'+"#"*200);

  def ProvidedLangs(self):
    i=1; print("\n"+"#"*10,"지원 원어","#"*50 )
    for lang in langs:
      print(f"{i:2}: {lang:10}",end="")
      if i%8==0: print();
      i += 1

  def TextLoad(self):                                           # 사용자가 선택하는 외국어 TEXT를 입력
    self.TEXT_DB = File().fileList()
    print("\n"+"=="*25)
    for key,val in self.TEXT_DB.items():
      print(f'{key}. {val.replace(".txt","")}')
    self.selectedInput = input(f"***** 불러올 TEXT를 선택하세요  >>> ")
    if self.selectedInput == '기타':
      textString = input(f"***** 읽어들일 문자열을 직접 입력하세요 >>> ")
    elif self.selectedInput in str(list(self.TEXT_DB.keys())) and self.selectedInput != '' :
      self.textTitle = self.TEXT_DB[int(self.selectedInput)]
      textString=File().fileRead(self.textTitle)
    return textString

  def TransString(self, textString):                            # 외국어 text에 대해 리스닝 학습을 위한 MP3 추출
    transObj = googletrans.Translator()
    self.ProvidedLangs()
    outText = int(input(f"\n***** 번역하고자 하는 언어를 선택하세요 (1-16 선택) >>> "))
    if 1 <= outText <= 16:
      outLang=langCodes[outText-1]
    else: print("\n***** 올바른 번역 언어를 선택하세요!")
    translatedString = transObj.translate(textString, dest=outLang, src='auto')
    transOut=translatedString.text
    detectedInLan=translatedString.src
    print("\n"*2,transOut)
    return textString, transOut, detectedInLan, outLang

  def TextToSpeech(self, textString, inLang):                   # 외국어 text에 대해 리스닝 학습을 위한 MP3 추출
    tts = gTTS(text=textString, lang=inLang)
    saveAudioYN = input(f"\n***** 생성된 오디오 파일을 저장할까요 (Y/N)? >>> ").lower()
    if saveAudioYN == "y":
      if self.selectedInput != '기타' :                             # 파일제목을 새로 입력하지 않아도 되는 기존파일을 선택했을 경우
        saveAudio=re.sub(r'.txt', '_'+inLang+'.mp3', self.textTitle)
      else :
        saveAudio=self.fileTitle+'_'+inLang+'.mp3'
      tts.save(saveAudio)
      print(f"생성된 오디오 파일이 {saveAudio}로 저장되었습니다.")
    #ps.playsound(outMP3)

  def VocaBook(self, textString,inLagn,outLang):                # 외국어 text에 대해 빈출 단어장 생성
    words_list=re.sub(r'[.,!#\n\';:"-]+', '', textString).lower()
    words_list = words_list.split()
    words_set=set(words_list); words_db={}
    for W in words_set:
      if W not in deletingEntry:
        words_db[W]=words_list.count(W)
    words_db2=sorted(words_db.items(), key=lambda x: -x[1])
    nomVocas = int(input(f"\n***** 생성할 단어 개수를 입력하세요 >>> "))
    if self.selectedInput != '기타' :
      saveVocas=re.sub(r'.txt', '_VOCA.txt', self.textTitle)
    else :
      saveVocas=self.fileTitle+'_VOCA.txt'
    transObj2 = googletrans.Translator()
    print("\n\n","#"*58,'\n',"#"*24,"단 어 장","#"*24, "\n","#"*58)
    f_voca = open(saveVocas,'w')
    vocaDB={}; i=1
    for ckey in words_db2:
      if i <= nomVocas:
        translatedVoca=transObj2.translate(ckey[0], dest=outLang, src=inLang).text
        f_voca.write(ckey[0]+"\t"*2+translatedVoca+'\n')
        print(f" {ckey[0]:<10}\t\t{translatedVoca:<10}")
        vocaDB[ckey[0]]=translatedVoca
      i += 1
    f_voca.close()
    return vocaDB

  def Quiz(self, vocaDB):                               # 외국어 text에 대해 빈출 단어장 생성 후 이를 기반으로 Quiz
    listKeys=list(vocaDB.keys())
    correct_no=0; wrong_no=0;
    i=1
    for voca in vocaDB:
      listKeys.remove(voca)
      sampling4=random.sample(listKeys,3)
      sampling4.append(voca); random.shuffle(sampling4)
      answer=sampling4.index(voca)+1
      print("\n ","#"*25,"\n ","#"*7,f"Quiz({i})번","#"*7,"\n ","#"*25,"\n ","="*80)
      reply=int(input(f"  {voca}:   (1)  {vocaDB[sampling4[0]]}   (2)  {vocaDB[sampling4[1]]}   (3)  {vocaDB[sampling4[2]]}   (4)  {vocaDB[sampling4[3]]} >> "))
      print(" ","="*80)
      listKeys.append(voca)
      if reply == answer:
        print(f"  정답입니다.\n"); correct_no += 1
      else: print(f"  오답입니다.\n"); wrong_no += 1
      i += 1
    print(f"  !!!! 수고하셨습니다. {correct_no+wrong_no}문항 중 {correct_no}문항 맞추셨습니다 !!!!!!!!!.\n")

  def FileSave(self,textString, postFix) :                        # 번역본 저장 여부 확인
    if self.selectedInput != '기타' :                             # 파일제목을 새로 입력하지 않아도 되는 기존파일을 선택했을 경우
      saveFilename=re.sub(r'.txt', '_'+postFix+'.txt', self.textTitle)
      File().fileDown(saveFilename,textString)
    else :                                                        # 파일제목을 입력해야하는 신규 파일일 경우
      self.fileTitle = input('파일 제목을 입력해주세요. >>> ')
      File().fileDown(self.fileTitle+'.txt',textString)
      self.TEXT_DB[len(self.TEXT_DB)+1] = self.fileTitle


class File :
  path = '/content/drive/MyDrive/00_KITA_2404/M2_Program/TEXTs/'
  outpath = '/content/drive/MyDrive/00_KITA_2404/M2_Program/'

  def fileRead(self, file) :                                      # 파일 Loading
    with open(self.path + file, 'r' ) as f :
      return f.read()

  def fileDown(self, fileTitle, fileText) :                       # File 저장: fileTitle-파일명, fileText-저장 내용
    try :
      with open(self.outpath + fileTitle,'w') as f:
        f.write(fileText + '\n')
        print('파일저장에 성공했습니다.')
    except Exception as e:
      print(f'####파일 저장에 실패하였습니다.####\n{e}')

  def fileList(self) :                                            # 디렉토리에 있는 파일 목록 가져오기
    listdir = sorted(os.listdir(self.path))
    inTextDict = {}
    for i in range(len(listdir)) :
      inTextDict[i+1] = listdir[i]
    return inTextDict

while True :
  obj1=TransSpeech()
  obj1.InitDisp()
  textString=obj1.TextLoad()
  textString, transOut,inLang, outLang=obj1.TransString(textString)

  saveYn = input('\n***** 번역된 내용을 저장하시겠습니까? (Y/N) >>> ').lower()
  if saveYn == 'y' :
    obj1.FileSave(transOut, outLang)

  genAudioYN = input(f"\n***** 음성 파일을 생성할까요 (Y/N)? >>> ").lower()
  if genAudioYN == "y":
    obj1.TextToSpeech(textString, inLang)

  genVocaYN = input(f"\n***** 입력 TEXT의 빈출 단어장을 생성할까요 (Y/N)? >>> ").lower()
  if genVocaYN == "y":
    vocaDB=obj1.VocaBook(textString,inLang,outLang)

  runQuizYN = input(f"\n***** Quiz를 시작할까요 (Y/N)? >>> ").lower()
  if runQuizYN == "y":
    obj1.Quiz(vocaDB)

  if input('\n***** 종료하시겠습니까? (Y/N) >>> ').lower() == 'y' :
    print('\n***** 수고하셨습니다. 프로그램을 종료합니다')
    break

```

Then, as the sun first appeared, we set off toward the mountains.
While we toiled up the steep path that wound around the moun-tain, I saw the abbey. I was amazed, not by the walls that girded it on every side, similar to others to be seen in all the Christian world, but by the bulk of what I later learned was the Aedificium. This was an octagonal construction that from a distance seemed a tetragon (a perfect form, which expresses the sturdiness and impregnability of the City of God), whose southern sides stood on the plateau of the abbey, while the northern ones seemed to grow from the steep side of the mountain, a sheer drop, to which they were bound. I might say that from below, at certain points, the cliff seemed to extend, reaching up toward the heavens, with the rock's same colors and mate-rial, which at a certain point became keep and tower (work of giants who had great familiarity with earth and sky). Three rows of windows proclaimed the triune rhythm of its elevation, so that what was physically squared on the earth was spiritually triangular in the sky. As we came closer, we realized that the quadrangular form included, at each of its corners, a heptagonal tower, five sides of which were visible on the outside —four of the eight sides, then, of the greater octagon producing four minor heptagons, which from the outside appeared as pentagons. And thus anyone can see the admirable concord of so many holy numbers, each revealing a subtle spiritual significance.
Eight, the number of perfection for every tetragon; four, the number of the Gospels; five, the number of the zones of the world; seven, the number of the gifts of the Holy Ghost. In its bulk and in its form, the Aedificium resembled Castel Ursino or Castel del Monte, which I was to see later in the south of the Italian peninsula, but its inaccessible position made it more awesome than those, and capable of inspiring fear in the traveler who approached it gradually. And it was fortunate that, since it was a very clear winter morning, I did not first see the building as it appears on stormy days.
I will not say, in any case, that it prompted feelings of jollity. I felt fear, and a subtle uneasiness. God knows these were not phantoms of my immature spirit, and I was rightly interpreting indubitable omens inscribed in the stone the day that the giants began their work, and before the deluded determination of the monks dared consecrate the building to the preservation of the divine word.
As our little mules strove up the last curve of the mountain, where the main path divided into three, producing two side paths, my master stopped for a while, to look around: at the sides of the road, at the road itself, and above the road, where, for a brief stretch, a series of evergreen pines formed a natural roof, white with snow.
